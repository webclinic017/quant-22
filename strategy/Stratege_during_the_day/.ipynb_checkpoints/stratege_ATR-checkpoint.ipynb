## ATR通道策略

### 策略思想
定义真实波动范围TR = max（abs（high - close_1）,abs(high - low),abs(low - close_1)）

计算TR的n日平均值ATR

取n1日均价为中轨，k1，k2为上下通道的系数

通道上轨 = n1日均价 + k1 * ATR

通道下轨 = n1日均价 - k2 * ATR

当价格突破上轨时，做多

当价格突破下轨时，做空

（止损止盈暂无，当由做多信号变为做空信号时，先平仓，在做空）

### 利用backtrade计算相关指标和通道,构建策略

```python
class ATR_Strategy(bt.Strategy):
    params = dict(
        N_ATR = 10, # ATR计算的周期
        N_avr = 10, # 均值计算的周期
        k_u = 2, # 上通道系数
        k_d = 2, # 下通道系数
        )
    
    def log(self, txt, dt=None):           
            dt = dt or self.datas[0].datetime.date(0)
            print('%s, %s' % (dt.isoformat(), txt))

    def __init__(self): 
        '''
        计算好相关指标
        '''
        ## 定义辅助指标
        self.order = None                   
        self.buy_count = 0 # 记录买入次数
        self.last_price = 0 # 记录买入价格
        self.buy_unit = 1 # 每次买卖一手
        
        ## 准备close、high、low 行情数据
        self.close = self.datas[0].close
        self.high = self.datas[0].high
        self.low = self.datas[0].low
        
        ## 计算策略所需指标
        # 计算atr值
#         self.atr = bt.talib.ATR(self.high, self.low, self.close, timeperiod=self.p.N_ATR)
        self.TR = bt.ind.Max((self.high(0)-self.low(0)), # 当日最高价-当日最低价
                                    abs(self.high(0)-self.close(-1)), # abs(当日最高价−前一日收盘价)
                                    abs(self.low(0)-self.close(-1))) # abs(当日最低价-前一日收盘价)
        self.atr = bt.ind.SimpleMovingAverage(self.TR, period=self.p.N_ATR, subplot=False)
        # 计算均价
        self.ave_close = bt.talib.SMA(self.close, timeperiod=self.p.N_avr)
        # 计算上下通道
        self.ATR_u = self.ave_close + self.p.k_u * self.atr
        self.ATR_d = self.ave_close - self.p.k_d * self.atr
        # 计算atr通道上下轨道突破信号
        self.crossover_u = bt.ind.CrossOver(self.close(0), self.ATR_u, subplot=False)
        self.crossover_d = bt.ind.CrossOver(self.close(0), self.ATR_d, subplot=False)
    
    
    def next(self): 
        '''
        策略逻辑：
            version:1
                如果持多仓，突破下轨时平仓并反向开空仓，
                如果持空仓，突破上轨时平仓并反向开多仓，
                如果未持仓，突破时开对应方向的仓位
                仓位为开1手，初始资金为20000
            
        '''
        ### version_1
        
        ## 如果还有订单在执行中，就不做新的仓位调整
        print(self.datas[0].datetime.date(0))
        print(self.position.size)
        if self.order:
            print('have order')
            return 
#         print(self.crossover_d[0])
        ## 当持有多单时
        if self.position.size > 0 :
            if self.crossover_d < 0:
                print("当前持有多单,且有空单信号")
                # 平仓并开空仓
                self.buy_unit = int(self.buy_unit) # 交易单位为手
                self.order = self.sell(self.datas[0],size=self.buy_unit * 2)
                self.last_price = self.position.price # 获取买入价格
                
            else:
                return
        ## 当持有空单时
        elif self.position.size < 0:
            if self.crossover_u > 0:
                print("当前持有空单,且有多单信号")
                # 平仓并开多仓
                self.buy_unit = int(self.buy_unit) # 交易单位为手
                self.order = self.buy(self.datas[0],size=self.buy_unit * 2)
                self.last_price = self.position.price # 获取买入价格
        ## 空仓时
        else:
            if self.crossover_u > 0:
                print("当前空仓,且有多单信号")
                # 平仓并开多仓
                self.buy_unit = int(self.buy_unit) # 交易单位为手
                self.order = self.buy(self.datas[0],size=self.buy_unit)
                self.last_price = self.position.price # 获取买入价格
            if self.crossover_d < 0:
                print("当前空仓,且有空单信号")
                # 平仓并开多仓
#                 self.buy_unit = int(self.buy_unit) # 交易单位为手
                self.order = self.sell(self.datas[0],size=self.buy_unit)
                self.last_price = self.position.price # 获取买入价格
        
    # 打印订单日志
    def notify_order(self, order):
        order_status = ['Created','Submitted','Accepted','Partial',
                        'Completed','Canceled','Expired','Margin','Rejected']
        # 未被处理的订单
        if order.status in [order.Submitted, order.Accepted]:
            self.log('ref:%.0f, name: %s, Order: %s'% (order.ref,
                                                   order.data._name,
                                                   order_status[order.status]))
            return
        # 已经处理的订单
        if order.status in [order.Partial, order.Completed]:
            if order.isbuy():
                self.log(
                        'BUY EXECUTED, status: %s, ref:%.0f, name: %s, Size: %.2f, Price: %.2f, Cost: %.2f, Comm %.2f' %
                        (order_status[order.status], # 订单状态
                         order.ref, # 订单编号
                         order.data._name, # 股票名称
                         order.executed.size, # 成交量
                         order.executed.price, # 成交价
                         order.executed.value, # 成交额
                         order.executed.comm)) # 佣金
            else: # Sell
                self.log('SELL EXECUTED, status: %s, ref:%.0f, name: %s, Size: %.2f, Price: %.2f, Cost: %.2f, Comm %.2f' %
                            (order_status[order.status],
                             order.ref,
                             order.data._name,
                             order.executed.size,
                             order.executed.price,
                             order.executed.value,
                             order.executed.comm))
                    
        elif order.status in [order.Canceled, order.Margin, order.Rejected, order.Expired]:
            # 订单未完成
            self.log('ref:%.0f, name: %s, status: %s'% (
                order.ref, order.data._name, order_status[order.status]))
            
        self.order = None
        
    def notify_trade(self, trade):
        # 交易刚打开时
        if trade.justopened:
            self.log('Trade Opened, name: %s, Size: %.2f,Price: %.2f' % (
                    trade.getdataname(), trade.size, trade.price))
        # 交易结束
        elif trade.isclosed:
            self.log('Trade Closed, name: %s, GROSS %.2f, NET %.2f, Comm %.2f' %(
            trade.getdataname(), trade.pnl, trade.pnlcomm, trade.commission))
        # 更新交易状态
        else:
            self.log('Trade Updated, name: %s, Size: %.2f,Price: %.2f' % (
                    trade.getdataname(), trade.size, trade.price))
   
  
```

```python
# 创建主控制器
cerebro = bt.Cerebro()

# 准备日线数据，输入到backtrader
datafeed = bt.feeds.PandasData(dataname=df,
                           fromdate=pd.to_datetime(f'{start_date}'),
                           todate=pd.to_datetime(f'{end_date}'))
cerebro.adddata(datafeed, name=f'{security}')

# 设置交易信息：初始资金 20000；
cerebro.broker.setcash(20000)
cerebro.broker.setcommission(commission=1.51, # 按 0.1% 来收取手续费
                             mult=10, # 合约乘数
                             margin=0.12, # 保证金比例
                             commtype=bt.CommInfoBase.COMM_FIXED,
                             stocklike=False)
cerebro.broker.set_slippage_fixed(fixed=0.5)

# 加入策略
cerebro.addstrategy(ATR_Strategy)
# 回测时需要添加 PyFolio 分析器
cerebro.addanalyzer(bt.analyzers.PyFolio, _name='pyfolio')
cerebro.addanalyzer(bt.analyzers.TimeReturn, _name='_TimeReturn')
result = cerebro.run()
# 借助 pyfolio 进一步做回测结果分析

pyfolio = result[0].analyzers.pyfolio # 注意：后面不要调用 .get_analysis() 方法
# 或者是 result[0].analyzers.getbyname('pyfolio')
returns, positions, transactions, gross_lev = pyfolio.get_pf_items()

```

```python
# 提取收益序列
pnl = pd.Series(result[0].analyzers._TimeReturn.get_analysis())
# 计算累计收益
cumulative = (pnl + 1).cumprod()
# 计算回撤序列
max_return = cumulative.cummax()
drawdown = (cumulative - max_return) / max_return
# 计算收益评价指标
import pyfolio as pf
# 按年统计收益指标
perf_stats_year = (pnl).groupby(pnl.index.to_period('y')).apply(lambda data: pf.timeseries.perf_stats(data)).unstack()
# 统计所有时间段的收益指标
perf_stats_all = pf.timeseries.perf_stats((pnl)).to_frame(name='all')
perf_stats = pd.concat([perf_stats_year, perf_stats_all.T], axis=0)
perf_stats_ = round(perf_stats,4).reset_index()


# 绘制图形
import matplotlib.pyplot as plt
plt.rcParams['axes.unicode_minus'] = False  # 用来正常显示负号
import matplotlib.ticker as ticker # 导入设置坐标轴的模块
# plt.style.use('seaborn') # plt.style.use('dark_background')
plt.style.use('dark_background')
fig, (ax0, ax1) = plt.subplots(2,1, gridspec_kw = {'height_ratios':[1.5, 4]}, figsize=(20,8))
cols_names = ['date', 'Annual_return', 'Cumulative\nreturns', 'Annual\nvolatility',
       'Sharpe\nratio', 'Calmar\nratio', 'Stability', 'Max\ndrawdown',
       'Omega\nratio', 'Sortino\nratio', 'Skew', 'Kurtosis', 'Tail\nratio',
       'Daily value\nat risk']

# 绘制表格
ax0.set_axis_off() # 除去坐标轴
table = ax0.table(cellText = perf_stats_.values,
                bbox=(0,0,1,1), # 设置表格位置， (x0, y0, width, height)
                rowLoc = 'right', # 行标题居中
                cellLoc='right' ,
                colLabels = cols_names, # 设置列标题
                colLoc = 'right', # 列标题居中
                edges = 'open' # 不显示表格边框
                )
table.set_fontsize(13)

# 绘制累计收益曲线
ax2 = ax1.twinx()
ax1.yaxis.set_ticks_position('right') # 将回撤曲线的 y 轴移至右侧
ax2.yaxis.set_ticks_position('left') # 将累计收益曲线的 y 轴移至左侧
# 绘制回撤曲线
drawdown.plot.area(ax=ax1, label='drawdown (right)', rot=0, alpha=0.3, fontsize=13, grid=False)
# 绘制累计收益曲线
(cumulative).plot(ax=ax2, color='#F1C40F' , lw=3.0, label='cumret (left)', rot=0, fontsize=13, grid=False)
# 不然 x 轴留有空白
ax2.set_xbound(lower=cumulative.index.min(), upper=cumulative.index.max())
# 主轴定位器：每 5 个月显示一个日期：根据具体天数来做排版
ax2.xaxis.set_major_locator(ticker.MultipleLocator(100))
# 同时绘制双轴的图例
h1,l1 = ax1.get_legend_handles_labels()
h2,l2 = ax2.get_legend_handles_labels()
plt.legend(h1+h2,l1+l2, fontsize=12, loc='upper left', ncol=1)

fig.tight_layout() # 规整排版
plt.show()
```

```python
# import base as b
import datetime
import numpy as np
import pandas as pd
# import tushare
import copy
# import plotly.plotly as py
# import plotly.graph_objs as go
# from plotly.offline import iplot,init_notebook_mode
# from base.datas.functions import *
import cufflinks

cufflinks.go_offline(connected=True)
import sys
# sys.path.append(r'..\..')
# sys.path.append(r'..\..')
# from base.utils import pdf
# from base.utils import get_trade_date
import trading_calendars as tc
import pprint


class Performance():
    '''输入单笔交易,日净利润, pandas  series 格式
    输出性能df'''

    def __init__(self, 单笔收益=None, 日收益=None, strat_allname=None, strat_name=None, 日累计收益=None, 初始资金=1000000, 组合策略数=1,
                 日交易次数=None, startTime=None, endTime=None, 单次滑点单位元=0, 日收益按交易日补缺=True, 随机采样率=1, tradelist=pd.DataFrame(),
                 *args, **kwargs):

        if isinstance(单笔收益, pd.DataFrame):
            self.单笔收益 = 单笔收益.净利润
            self.单笔收益trade_df = 单笔收益
        else:
            self.单笔收益 = 单笔收益
        if isinstance(self.单笔收益, pd.Series):
            self.tradedf = self.单笔收益
            self.tradedf = self.tradedf.to_frame(name="净利润")

        if isinstance(日收益, pd.DataFrame):
            self.日收益 = 日收益.净利润
        else:
            self.日收益 = 日收益
        if isinstance(self.日收益, pd.Series):
            self.tradedf = self.日收益
            self.tradedf = self.tradedf.to_frame(name="净利润")

        self.随机采样率 = 随机采样率

        self.单笔pe = None
        self.日pe = None

        self.日累计收益 = 日累计收益

        self.组合策略数 = 组合策略数
        self.初始资金 = 初始资金 * 组合策略数
        self.日交易次数 = 日交易次数
        self.startTime = startTime
        self.endTime = endTime
        if not tradelist.empty:
            self.tradelist = tradelist[self.startTime:self.endTime]

        # 读取数据库rolling所有的参数及收益
        try:
            self.parameters = rolling_optuni_results_from_mongo(
                collection=self.tradedf.strat_allname, dbname='rolling_backtest_strategy_pars')[0]
        except Exception as e:
            pass
            # print(e)
            self.parameters = None
        self.单次滑点单位元 = 单次滑点单位元
        self.strat_name = strat_name
        self.strat_allname = strat_allname

        if type(self.单笔收益) == pd.core.series.Series:
            self.单笔收益 = self.单笔收益[startTime:endTime]  # .sample(frac=随机采样率)
            self.单笔收益.sort_index(inplace=True)
            self._单笔收益回撤(self.单笔收益)
            self._日收益回撤(self.单笔收益回撤df.净利润, 日收益按交易日补缺=日收益按交易日补缺)
            self.基于单笔收益计算性能(self.单笔收益回撤df.净利润)
            self.基于日收益计算性能(self.日收益回撤df.日净利润, 日收益按交易日补缺=日收益按交易日补缺)
        elif type(self.日收益) == pd.core.series.Series:
            self.日收益 = self.日收益[startTime:endTime]
            self._日收益回撤(日收益=self.日收益, 日收益按交易日补缺=日收益按交易日补缺)
            self.基于日收益计算性能(self.日收益, 日收益按交易日补缺=日收益按交易日补缺)
            self.单笔pe = pd.Series()
        elif type(self.日累计收益) == pd.core.series.Series:
            self.日收益 = (self.日累计收益 - self.日累计收益.shift(1)) * 10000
            # self.日收益 = self.日收益.sample(frac=随机采样率)
            self._日收益回撤(日收益=self.日收益, 日收益按交易日补缺=日收益按交易日补缺)
            self.基于日收益计算性能(self.日收益, 日收益按交易日补缺=日收益按交易日补缺)
            self.单笔pe = pd.Series()

        else:
            raise Exception('请传入收益序列,格式为 pandas series[要求数据列名为"日期"且为datatime64格式]')

        self.日pe_c1 = [
            # '策略名',
            # '策略全名',
            # '起始时间', '结束时间', '初始资金', '期末资金',
            '初始资金年化收益',
            # '最大连续盈利天数为',
            # '最大连续亏损天数为',
            '总盈亏',
            '最大回撤金额为',
            '初始资金最大回撤率为',
            # '当前回撤金额为',
            '初始资金当前回撤率为',
            # '最大回撤率开始日期', '最大回撤率结束日期',
            #      'sharpe_ratio_v0',
            #      'sharpe_ratio_v1',
            'sharpe_ratio_v4',
            # 'calmar_ratio', '盈利天数',
            # '亏损天数', '日胜率', '周胜率', '月胜率', '最大日亏损', '最大日盈利', '日平均收益', '日平均收益率百分之',
            # '亏损日平均亏损', '盈利日平均收益', '日盈亏比'
        ]
        self.日pe_c2 = [
            # '策略名',
            # '策略全名',
            # '起始时间', '结束时间', '初始资金', '期末资金',
            '初始资金年化收益',
            # '最大连续盈利天数为',
            # '最大连续亏损天数为',
            '总盈亏',
            # '最大回撤金额为',
            '初始资金最大回撤率为',
            # '当前回撤金额为',
            # '初始资金当前回撤率为',
            # '最大回撤率开始日期', '最大回撤率结束日期',
            #      'sharpe_ratio_v0',
            #      'sharpe_ratio_v1',
            'sharpe_ratio_v4',
            # 'calmar_ratio', '盈利天数',
            # '亏损天数', '日胜率', '周胜率', '月胜率', '最大日亏损', '最大日盈利', '日平均收益', '日平均收益率百分之',
            # '亏损日平均亏损', '盈利日平均收益', '日盈亏比'
        ]
        self.单笔pe_c1 = [
            # '亏损单笔平均亏损', '盈利单笔平均收益', '最大单笔亏损', '最大单笔盈利',
            '总交易笔数',

            '交易日日平均交易笔数',
            # '盈利笔数',
            # '亏损笔数', '单笔胜率', '单笔盈亏比',
            '单笔平均收益',
            '总滑点金额',
            # '单笔平均收益率百分之'
        ]
        self.单笔pe_c2 = [
            # '亏损单笔平均亏损', '盈利单笔平均收益', '最大单笔亏损', '最大单笔盈利',
            '总交易笔数',

            # '交易日日平均交易笔数',
            # '盈利笔数',
            # '亏损笔数', '单笔胜率', '单笔盈亏比',
            # '单笔平均收益',
            '总滑点金额',
            # '单笔平均收益率百分之'
        ]

    def period(self, date_from='2019-03-01', date_to=None):

        if self.日收益回撤df.index[-1] > datetime.datetime.strptime(date_from, '%Y-%m-%d'):
            (self.日收益回撤df.日净利润[date_from:date_to].cumsum() / 10000).iplot(title=f'{date_from}:{date_to}盈利',
                                                                          yTitle='单位:万',
                                                                          colors='blue', )  # annotations={'2015-09-01':'股指开始限仓'}
            (self.日收益回撤df.日累计最大回撤金额[date_from:date_to] / 10000).iplot(title=f'{date_from}:{date_to}回撤', yTitle='单位：万',
                                                                      colors='blue')

    def p3(self, **kwargs):
        self.简要pe()
        时期 = f"{self.日收益回撤df.index[0].date()} 至 {self.日收益回撤df.index[-1].date()}"

        (self.日收益回撤df[['日累计净利润', '日累计最大回撤金额']] / 10000).iplot(**kwargs, subplots=True, shared_xaxes=True,
                                                              shared_yaxes=False,
                                                              dimensions=(1500, 600),
                                                              shape=(4, 1),
                                                              specs=[[{'rowspan': 3}], [None], [None], [{}], ],
                                                              subplot_titles=['总盈利', '回撤'],

                                                              title=时期, yTitle='单位:万',
                                                              colors='blue', )  # annotations={'2015-09-01':'股指开始限仓'}
        (self.日收益回撤df[['日收益率累加', '日累计最大回撤比率']]).iplot(**kwargs, subplots=True, shared_xaxes=True, shared_yaxes=False,
                                                      dimensions=(1500, 600),
                                                      shape=(4, 1),
                                                      specs=[[{'rowspan': 3}], [None], [None], [{}]],
                                                      subplot_titles=['日收益率累加(复利)', '累计回撤率'],

                                                      title=时期, yTitle='单位:百分比',
                                                      colors='blue', )  # annotations={'2015-09-01':'股指开始限仓'}
        (self.日收益回撤df[['累计净利润初始资金收益率', '日初始资金最大回撤比率']]).iplot(**kwargs, subplots=True, shared_xaxes=True, shared_yaxes=False,
                                                        dimensions=(1500, 600),
                                                        shape=(4, 1),
                                                        specs=[[{'rowspan': 3}], [None], [None], [{}]],
                                                        subplot_titles=['总收益率(单利)', '相对初始资金最大回撤率'],

                                                        title=时期, yTitle='单位:百分比',
                                                        colors='blue', )  # annotations={'2015-09-01':'股指开始限仓'}

        if type(self.单笔收益) == pd.core.series.Series:
            return pdf(pd.concat([pd.DataFrame(self.日pe), pd.DataFrame(self.单笔pe)]), 200)
        else:
            return pdf(pd.DataFrame(self.日pe), 200)

    def p1(self):

        self.简要pe()
        时期 = f"{self.日收益回撤df.index[0].date()} 至 {self.日收益回撤df.index[-1].date()}"

        (self.日收益回撤df[['日累计净利润', '日累计最大回撤金额']] / 10000).iplot(subplots=True, shared_xaxes=True, shared_yaxes=False,
                                                              # dimensions=(1000, 600),
                                                              shape=(4, 1),
                                                              specs=[[{'rowspan': 3}], [None], [None], [{}], ],
                                                              subplot_titles=['总盈利', '回撤'],

                                                              title=时期, yTitle='单位:万',
                                                              colors='blue', )  # annotations={'2015-09-01':'股指开始限仓'}
        # (self.日收益回撤df.日累计最大回撤金额 / 10000).iplot(title='回撤', yTitle='单位：万', colors='blue')
        df_year = self.日收益回撤df.groupby(pd.Grouper(freq="1Y")).cumsum()
        (df_year['日净利润'] / 10000).iplot(title='年度盈亏', yTitle='单位：万', colors='blue', dimensions=(1000, 300))

        if '当日交易次数' in self.日收益回撤df.columns:
            (self.日收益回撤df.当日交易次数).iplot(kind='bar', title='日交易次数', yTitle='单位:开平各算1次', dimensions=(1000, 300),
                                        colors='blue', )  # annotations={'2015-09-01':'股指开始限仓'}
            (self.日收益回撤df.当日交易次数).iplot(dimensions=(1000, 300), kind='histogram', title='日交易次数分布', yTitle='单位:开平各算1次',
                                        colors='blue', )  # annotations={'2015-09-01':'股指开始限仓'}

        (self.日收益回撤df.日净利润 / 10000).iplot(dimensions=(1000, 300), kind='histogram', title='日盈亏分布', yTitle='累计次数',
                                          xTitle='单位：万', colors='blue')

        df_day_mean = self.日收益回撤df.groupby(by=self.日收益回撤df.index.day).mean()
        (df_day_mean['日净利润'] / 10000).iplot(dimensions=(1000, 300), kind='bar', title='所有月份日平均盈亏', yTitle='单位：万',
                                            colors='blue')

        (self.日收益回撤df.日净利润 / 10000).iplot(dimensions=(1000, 300), kind='bar', title='日盈亏', yTitle='单位：万', colors='blue')

        (self.日收益回撤df.日净利润 / 10000).iplot(dimensions=(1000, 300), kind='scatter', title='日盈亏', yTitle='单位：万',
                                          mode='markers', size=5,
                                          colors='blue')

        self.日收益回撤df_week = self.日收益回撤df.resample('W').sum()

        (self.日收益回撤df_week['日净利润'] / 10000).iplot(dimensions=(1000, 300), kind='bar', title='周盈亏', yTitle='单位：万',
                                                  colors='blue')

        self.日收益回撤df_month = self.日收益回撤df.resample('M').sum()

        (self.日收益回撤df_month['日净利润'] / 10000).iplot(dimensions=(1000, 300), kind='bar', title='月盈亏', yTitle='单位：万',
                                                   colors='blue')

        df_month_mean = self.日收益回撤df_month.groupby(by=self.日收益回撤df_month.index.month).mean()
        (df_month_mean['日净利润'] / 10000).iplot(dimensions=(1000, 300), kind='bar', title='所有年份月度平均盈亏', yTitle='单位：万',
                                              colors='blue')

        # todo:
        df_day_mean = self.日收益回撤df.groupby(by=self.日收益回撤df.index.dayofweek).sum()
        (df_day_mean['日净利润'] / 10000).iplot(dimensions=(1000, 300), kind='bar', title='星期总盈亏', yTitle='单位：万',
                                            colors='blue')

        df_year = self.日收益回撤df.resample('Y').sum()

        df_year.index = df_year.index.to_period()
        (df_year['日净利润'] / 10000).iplot(dimensions=(600, 300), kind='bar', title='年度盈亏', yTitle='单位：万', colors='blue')

        if type(self.单笔收益) == pd.core.series.Series:
            return pdf(pd.concat([pd.DataFrame(self.日pe), pd.DataFrame(self.单笔pe)]), 200)
        else:
            return pdf(pd.DataFrame(self.日pe), 200)

    def p2(self):
        self.简要pe()
        # (self.单笔收益回撤df.净利润['2019-03-01':].cumsum() / 10000).iplot(title='190301盈利', yTitle='单位:万',
        #                                                           colors='blue', )  # annotations={'2015-09-01':'股指开始限仓'}
        # (self.单笔收益回撤df.累计最大回撤金额['2019-03-01':] / 10000).iplot(title='回撤', yTitle='单位：万', colors='blue')

        df_year = self.单笔收益回撤df.groupby(pd.Grouper(freq="1Y")).cumsum()
        (df_year['净利润'] / 10000).iplot(title='年度盈亏', yTitle='单位：万', colors='blue')

        (self.单笔收益回撤df.累计净利润 / 10000).iplot(title='总盈利', yTitle='单位:万',
                                            colors='blue', )  # annotations={'2015-09-01':'股指开始限仓'}
        (self.单笔收益回撤df.净利润 / 10000).iplot(kind='histogram', title='单笔盈亏分布', yTitle='累计次数', xTitle='单位：万', colors='blue')

        (self.单笔收益回撤df.净利润 / 10000).iplot(kind='bar', title='单笔盈亏', yTitle='单位：万', colors='blue')

        (self.单笔收益回撤df.净利润 / 10000).iplot(kind='scatter', title='单笔盈亏', yTitle='单位：万', mode='markers', size=5,
                                          colors='blue')
        (self.单笔收益回撤df.累计最大回撤金额 / 10000).iplot(title='回撤', yTitle='单位：万', colors='blue')

        df_week = self.单笔收益回撤df.resample('W').sum()

        (df_week['净利润'] / 10000).iplot(kind='bar', title='周盈亏', yTitle='单位：万', colors='blue')

        df_month = self.单笔收益回撤df.resample('M').sum()

        (df_month['净利润'] / 10000).iplot(kind='bar', title='月盈亏', yTitle='单位：万', colors='blue')

        df_year = self.单笔收益回撤df.resample('Y', ).sum()

        df_year.index = df_year.index.to_period()
        (df_year['净利润'] / 10000).iplot(kind='bar', title='年度盈亏', yTitle='单位：万', colors='blue')

        return self.详细pe()
        # return pdf(pd.DataFrame(self.单笔pe),200)

    # def _单笔收益回撤(self, 单笔收益):
    #     df_cal = pd.DataFrame()
    #
    #     if self.单次滑点单位元:
    #         df_cal['净利润'] = 单笔收益 - (self.单次滑点单位元 * 2)
    #     else:
    #         df_cal['净利润'] = 单笔收益
    #
    #     df_cal['当笔日交易次数'] = df_cal.groupby('日期').cumcount() * 2 + 2
    #     if self.日交易次数:
    #         df_cal = df_cal[df_cal.当笔日交易次数 <= self.日交易次数]
    #     # print(df_cal.净利润.sum())
    #     # df_cal = df_cal.sample(frac=self.随机采样率)
    #     df_cal['累计净利润'] = df_cal['净利润'].cumsum()
    #     df_cal['权益'] = df_cal['累计净利润'] + self.初始资金
    #     df_cal['期初权益'] = df_cal['累计净利润'] - df_cal['净利润'] + self.初始资金
    #     df_cal['日累计收益率'] = df_cal['累计净利润'] * 100 / self.初始资金
    #     df_cal['收益率'] = (df_cal['净利润'] / df_cal['权益'].shift(1))
    #     df_cal['日最大权益'] = df_cal['权益'].expanding(min_periods=1).max()  # 计算之前的账户最大价值
    #     df_cal['累计最大回撤金额'] = df_cal['权益'] - df_cal['日最大权益']
    #     df_cal['累计最大回撤比率'] = df_cal['权益'] / df_cal['日最大权益'] - 1  # 计算回撤
    #
    #     self.单笔收益回撤df = df_cal

    def get_net_value(self, daily_return):
        _net_value = (daily_return + 1).cumprod()  # 这时候第一个值还不是1
        # 因为第一个是NaN值
        net_value = _net_value / _net_value.iloc[0]
        return net_value

    def _单笔收益回撤(self, 单笔收益):
        df_cal = pd.DataFrame()

        if self.单次滑点单位元:
            df_cal['净利润'] = 单笔收益 - (self.单次滑点单位元 * 2)
        else:
            df_cal['净利润'] = 单笔收益

        df_cal['当笔日交易次数'] = df_cal.groupby('日期').cumcount() * 2 + 2
        if self.日交易次数:
            df_cal = df_cal[df_cal.当笔日交易次数 <= self.日交易次数]
        # print(df_cal.净利润.sum())
        # df_cal = df_cal.sample(frac=self.随机采样率)
        df_cal['累计净利润'] = df_cal['净利润'].cumsum()
        df_cal['权益'] = df_cal['累计净利润'] + self.初始资金
        df_cal['期初权益'] = df_cal['累计净利润'] - df_cal['净利润'] + self.初始资金
        df_cal['累计净利润初始资金收益率'] = df_cal['累计净利润'] * 100 / self.初始资金
        # df_cal['日收益率(单利)'] = (df_cal['净利润'] / df_cal['权益'].shift(1))  # 单利
        df_cal['日收益率(单利)'] = (df_cal['净利润'] / df_cal['期初权益'])  # 单利
        df_cal['日收益率(复利)'] = (df_cal['净利润'] * (df_cal['期初权益'] / self.初始资金) / df_cal['期初权益'])  # 复利   = df_cal['净利润']/self.初始资金
        df_cal['单利净值'] = self.get_net_value(df_cal['日收益率(单利)'])
        df_cal['复利净值'] = self.get_net_value(df_cal['日收益率(复利)'])
        df_cal['日最大权益'] = df_cal['权益'].expanding(min_periods=1).max()  # 计算之前的账户最大价值
        df_cal['累计最大回撤金额'] = df_cal['权益'] - df_cal['日最大权益']
        df_cal['累计最大回撤比率'] = df_cal['权益'] / df_cal['日最大权益'] - 1  # 计算回撤

        self.单笔收益回撤df = df_cal
    def p4(self):
        self.简要pe()
        时期 = f"{self.日收益回撤df.index[0].date()} 至 {self.日收益回撤df.index[-1].date()}"

        (self.日收益回撤df[['日累计净利润', '日累计最大回撤金额']] / 10000).iplot(subplots=True, shared_xaxes=True, shared_yaxes=False,
                                                              # dimensions=(1000, 600),
                                                              shape=(4, 1),
                                                              specs=[[{'rowspan': 3}], [None], [None], [{}], ],
                                                              subplot_titles=['总盈利', '回撤'],

                                                              title=时期, yTitle='单位:万',
                                                              colors='blue', )  # annotations={'2015-09-01':'股指开始限仓'}
        # (self.日收益回撤df.日累计最大回撤金额 / 10000).iplot(title='回撤', yTitle='单位：万', colors='blue')
        (self.日收益回撤df[['累计净利润初始资金收益率', '日初始资金最大回撤比率']]).iplot(subplots=True, shared_xaxes=True, shared_yaxes=False,
                                                        # dimensions=(1500, 600),
                                                        shape=(4, 1),
                                                        specs=[[{'rowspan': 3}], [None], [None], [{}]],
                                                        subplot_titles=['总收益率(单利)', '相对初始资金最大回撤率'],

                                                        title=时期, yTitle='单位:百分比',
                                                        colors='blue', )  # annotations={'2015-09-01':'股指开始限仓'}

        df_year = self.日收益回撤df.groupby(pd.Grouper(freq="1Y")).cumsum()
        (df_year['日净利润'] / 10000).iplot(title='年度盈亏', yTitle='单位：万', colors='blue', )  # dimensions=(1000, 300))

        if '当日交易次数' in self.日收益回撤df.columns:
            (self.日收益回撤df.当日交易次数).iplot(kind='bar', title='日交易次数', yTitle='单位:开平各算1次', dimensions=(1000, 300),
                                        colors='blue', )  # annotations={'2015-09-01':'股指开始限仓'}
            # (self.日收益回撤df.当日交易次数).iplot(dimensions=(1000, 300), kind='histogram', title='日交易次数分布', yTitle='单位:开平各算1次',
            #                             colors='blue', )  # annotations={'2015-09-01':'股指开始限仓'}

        df_day_mean = self.日收益回撤df.groupby(by=self.日收益回撤df.index.day).mean()
        # (df_day_mean['日净利润'] / 10000).iplot(dimensions=(1000, 300), kind='bar', title='所有月份日平均盈亏', yTitle='单位：万',
        #                                     colors='blue')

        (self.日收益回撤df.日净利润 / 10000).iplot(dimensions=(1000, 300), kind='bar', title='日盈亏', yTitle='单位：万', colors='blue')

        # (self.日收益回撤df.日净利润 / 10000).iplot(dimensions=(1000, 300), kind='scatter', title='日盈亏', yTitle='单位：万',
        #                                   mode='markers', size=5,
        #                                   colors='blue')

        self.日收益回撤df_week = self.日收益回撤df.resample('W').sum()

        (self.日收益回撤df_week['日净利润'] / 10000).iplot(dimensions=(1000, 300), kind='bar', title='周盈亏', yTitle='单位：万',
                                                  colors='blue')

        self.日收益回撤df_month = self.日收益回撤df.resample('M').sum()

        (self.日收益回撤df_month['日净利润'] / 10000).iplot(dimensions=(1000, 300), kind='bar', title='月盈亏', yTitle='单位：万',
                                                   colors='blue')

        # df_month_mean = self.日收益回撤df_month.groupby(by=self.日收益回撤df_month.index.month).mean()
        # (df_month_mean['日净利润'] / 10000).iplot(dimensions=(1000, 300), kind='bar', title='所有年份月度平均盈亏', yTitle='单位：万',
        #                                       colors='blue')

        df_year = self.日收益回撤df.resample('Y').sum()

        df_year.index = df_year.index.to_period()
        (df_year['日净利润'] / 10000).iplot(dimensions=(600, 300), kind='bar', title='年度盈亏', yTitle='单位：万', colors='blue')

        (self.日收益回撤df.日净利润 / 10000).iplot(dimensions=(1000, 300), kind='histogram', title='日盈亏分布', yTitle='累计次数',
                                          xTitle='单位：万', colors='blue')

        (self.日收益回撤df_week.日净利润 / 10000).iplot(dimensions=(1000, 300), kind='histogram', title='周盈亏分布', yTitle='累计次数',
                                               xTitle='单位：万', colors='blue')

        (self.日收益回撤df_month.日净利润 / 10000).iplot(dimensions=(1000, 300), kind='histogram', title='月盈亏分布', yTitle='累计次数',
                                                xTitle='单位：万', colors='blue')

        return self.详细pe()

        # if type(self.单笔收益) == pd.core.series.Series:
        #     return pdf(pd.concat([pd.DataFrame(self.日pe), pd.DataFrame(self.单笔pe)]), 200,float=3)
        # else:
        #     return pdf(pd.DataFrame(self.日pe), 200,float=3)

    def p5(self, h=800, w=300):
        self.简要pe()

        时期 = f"{self.日收益回撤df.index[0].date()} 至 {self.日收益回撤df.index[-1].date()}"

        (self.日收益回撤df[['日累计净利润', '日累计最大回撤金额']] / 10000).iplot(subplots=True, shared_xaxes=True, shared_yaxes=False,
                                                              dimensions=(h, w),
                                                              shape=(4, 1),
                                                              specs=[[{'rowspan': 3}], [None], [None], [{}], ],
                                                              subplot_titles=['总盈利', '回撤'],

                                                              title=时期, yTitle='单位:万',
                                                              colors='blue', )  # annotations={'2015-09-01':'股指开始限仓'}
        # (self.日收益回撤df.日累计最大回撤金额 / 10000).iplot(title='回撤', yTitle='单位：万', colors='blue')
        # (self.日收益回撤df[['累计净利润初始资金收益率', '日初始资金最大回撤比率']]).iplot(subplots=True, shared_xaxes=True, shared_yaxes=False,
        #                                                 # dimensions=(1500, 600),
        #                                                 shape=(4, 1),
        #                                                 specs=[[{'rowspan': 3}], [None], [None], [{}]],
        #                                                 subplot_titles=['总收益率(单利)', '相对初始资金最大回撤率'],
        #
        #                                                 title=时期, yTitle='单位:百分比',
        #                                                 colors='blue', )  # annotations={'2015-09-01':'股指开始限仓'}
        #
        # df_year = self.日收益回撤df.groupby(pd.Grouper(freq="1Y")).cumsum()
        # (df_year['日净利润'] / 10000).iplot(title='年度盈亏', yTitle='单位：万', colors='blue',) #dimensions=(1000, 300))

        if '当日交易次数' in self.日收益回撤df.columns:
            (self.日收益回撤df.当日交易次数).iplot(kind='bar', title='日交易次数', yTitle='单位:开平各算1次', dimensions=(600, 200),
                                        colors='blue', )  # annotations={'2015-09-01':'股指开始限仓'}
            # (self.日收益回撤df.当日交易次数).iplot(dimensions=(1000, 100), kind='histogram', title='日交易次数分布', yTitle='单位:开平各算1次',
            #                             colors='blue', )  # annotations={'2015-09-01':'股指开始限仓'}

        #
        # df_day_mean = self.日收益回撤df.groupby(by=self.日收益回撤df.index.day).mean()
        # (df_day_mean['日净利润'] / 10000).iplot(dimensions=(1000, 300), kind='bar', title='所有月份日平均盈亏', yTitle='单位：万',
        #                                     colors='blue')

        # (self.日收益回撤df.日净利润 / 10000).iplot(dimensions=(1000, 300), kind='bar', title='日盈亏', yTitle='单位：万', colors='blue')

        # (self.日收益回撤df.日净利润 / 10000).iplot(dimensions=(1000, 300), kind='scatter', title='日盈亏', yTitle='单位：万',
        #                                   mode='markers', size=5,
        #                                   colors='blue')

        # self.日收益回撤df_week = self.日收益回撤df.resample('W').sum()
        #
        # (self.日收益回撤df_week['日净利润'] / 10000).iplot(dimensions=(1000, 300), kind='bar', title='周盈亏', yTitle='单位：万',
        #                                           colors='blue')
        #
        # self.日收益回撤df_month = self.日收益回撤df.resample('M').sum()
        #
        # (self.日收益回撤df_month['日净利润'] / 10000).iplot(dimensions=(1000, 300), kind='bar', title='月盈亏', yTitle='单位：万',
        #                                            colors='blue')

        # df_month_mean = self.日收益回撤df_month.groupby(by=self.日收益回撤df_month.index.month).mean()
        # (df_month_mean['日净利润'] / 10000).iplot(dimensions=(1000, 300), kind='bar', title='所有年份月度平均盈亏', yTitle='单位：万',
        #                                       colors='blue')

        # df_year = self.日收益回撤df.resample('Y').sum()
        #
        # df_year.index = df_year.index.to_period()
        # (df_year['日净利润'] / 10000).iplot(dimensions=(600, 300), kind='bar', title='年度盈亏', yTitle='单位：万', colors='blue')
        #
        # (self.日收益回撤df.日净利润 / 10000).iplot(dimensions=(1000, 300), kind='histogram', title='日盈亏分布', yTitle='累计次数',
        #                                   xTitle='单位：万', colors='blue')
        #
        # (self.日收益回撤df_week.日净利润 / 10000).iplot(dimensions=(1000, 300), kind='histogram', title='周盈亏分布', yTitle='累计次数',
        #                                   xTitle='单位：万', colors='blue')
        #
        # (self.日收益回撤df_month.日净利润 / 10000).iplot(dimensions=(1000, 300), kind='histogram', title='月盈亏分布', yTitle='累计次数',
        #                                   xTitle='单位：万', colors='blue')
        if type(self.单笔收益) == pd.core.series.Series:
            return pdf(pd.concat([pd.DataFrame(self.日pe), pd.DataFrame(self.单笔pe)]), 200, float=3)
        else:
            return pdf(pd.DataFrame(self.日pe), 200, float=3)

    def 详细pe(self):
        if type(self.单笔收益) == pd.core.series.Series:
            p = pdf(pd.concat([pd.DataFrame(self.日pe), pd.DataFrame(self.单笔pe)]), 200, float=3)
        else:
            p = pdf(pd.DataFrame(self.日pe), 200, float=3)
        p = p.reset_index()
        l = 10
        p1 = p[:l]
        for s in range(0, len(p) - l, l):
            # p1[str(s)]=p[s+l:s+l+l]['index'].values
            p1 = pd.concat([p1, p[s + l:s + l + l].reset_index(drop=True)], axis=1, ignore_index=True)
        p1.index = [''] * len(p1)
        p1.columns = [''] * p1.shape[1]
        return p1


    def 详细pe_v1(self):
        if type(self.单笔收益) == pd.core.series.Series:
            p = pdf(pd.concat([pd.DataFrame(self.日pe), pd.DataFrame(self.单笔pe)]), 200, float=3)
        else:
            p = pdf(pd.DataFrame(self.日pe), 200, float=3)
        p = p.reset_index()
        l = 10
        p1 = p[:l]
        # for s in range(0, len(p) - l, l):
        #     # p1[str(s)]=p[s+l:s+l+l]['index'].values
        #     p1 = pd.concat([p1, p[s + l:s + l + l].reset_index(drop=True)], axis=1, ignore_index=True)
        # p1.index = [''] * len(p1)
        # p1.columns = [''] * p1.shape[1]
        return p


    def 简要pe_v3(self, name=None):
        f = 3
        pd.options.display.float_format = '{:,.3f}'.replace('3f', f'{f}f').format
        # if isinstance(self.单笔pe, (pd.DataFrame, pd.Series)):
        # if not getattr(self, '单笔pe').empty:
        # print(self.单笔pe[self.单笔pe_c1])
        # if isinstance(self.日pe, (pd.DataFrame, pd.Series)):
        # if not getattr(self, '日pe').empty:
        # print(self.日pe[self.日pe_c1])

        if isinstance(self.单笔pe, (pd.DataFrame, pd.Series)):
            if not getattr(self, '单笔pe').empty:
                # return pd.concat([pd.DataFrame(self.日pe[self.日pe_c1]), pd.DataFrame(self.单笔pe[self.单笔pe_c1])])[0].T
                return \
                    pdf(pd.concat([pd.DataFrame(self.日pe[self.日pe_c2]), pd.DataFrame(self.单笔pe[self.单笔pe_c2])]),
                        float=2)[
                        0].apply(lambda x: round(x, 2) if isinstance(x, float) else x).to_frame(
                        name=name if name else self.strat_name).T
                # value_list = list(pdf(pd.concat([pd.DataFrame(self.日pe[self.日pe_c1]), pd.DataFrame(self.单笔pe[self.单笔pe_c1])]),
                #          float=2)[0].apply(lambda x: round(x, 2) if isinstance(x, float) else x))
                # name_list = self.日pe_c1
                # info_list = [":".join([name_list[index], str(value_list[index])]) for index in range(0, len(name_list))]
                # print(info_list)
                # pass

        else:
            return pdf(pd.DataFrame(self.日pe[self.日pe_c2]), 200, float=3)[0].apply(
                lambda x: round(x, 2) if isinstance(x, float) else x).to_frame(name=name if name else self.strat_name).T

    def 简要pe_v2(self):
        f = 3
        pd.options.display.float_format = '{:,.3f}'.replace('3f', f'{f}f').format
        # if isinstance(self.单笔pe, (pd.DataFrame, pd.Series)):
        # if not getattr(self, '单笔pe').empty:
        # print(self.单笔pe[self.单笔pe_c1])
        # if isinstance(self.日pe, (pd.DataFrame, pd.Series)):
        # if not getattr(self, '日pe').empty:
        # print(self.日pe[self.日pe_c1])

        if isinstance(self.单笔pe, (pd.DataFrame, pd.Series)):
            if not getattr(self, '单笔pe').empty:
                # return pd.concat([pd.DataFrame(self.日pe[self.日pe_c1]), pd.DataFrame(self.单笔pe[self.单笔pe_c1])])[0].T
                print((pdf(pd.concat([pd.DataFrame(self.日pe[self.日pe_c2]), pd.DataFrame(self.单笔pe[self.单笔pe_c2])]),
                           float=2)[0].apply(lambda x: round(x, 2) if isinstance(x, float) else x)).to_dict())
                # value_list = list(pdf(pd.concat([pd.DataFrame(self.日pe[self.日pe_c1]), pd.DataFrame(self.单笔pe[self.单笔pe_c1])]),
                #          float=2)[0].apply(lambda x: round(x, 2) if isinstance(x, float) else x))
                # name_list = self.日pe_c1
                # info_list = [":".join([name_list[index], str(value_list[index])]) for index in range(0, len(name_list))]
                # print(info_list)
                # pass

        else:
            print((pdf(pd.DataFrame(self.日pe[self.日pe_c2]), 200, float=3)[0].apply(
                lambda x: round(x, 2) if isinstance(x, float) else x)).to_dict())

    def 简要pe(self):
        f = 3
        pd.options.display.float_format = '{:,.3f}'.replace('3f', f'{f}f').format
        # if isinstance(self.单笔pe, (pd.DataFrame, pd.Series)):
        # if not getattr(self, '单笔pe').empty:
        # print(self.单笔pe[self.单笔pe_c1])
        # if isinstance(self.日pe, (pd.DataFrame, pd.Series)):
        # if not getattr(self, '日pe').empty:
        # print(self.日pe[self.日pe_c1])

        if isinstance(self.单笔pe, (pd.DataFrame, pd.Series)):
            if not getattr(self, '单笔pe').empty:
                # return pd.concat([pd.DataFrame(self.日pe[self.日pe_c1]), pd.DataFrame(self.单笔pe[self.单笔pe_c1])])[0].T
                # print((pdf(pd.concat([pd.DataFrame(self.日pe[self.日pe_c1]), pd.DataFrame(self.单笔pe[self.单笔pe_c1])]),float=2)[0].apply(lambda x: round(x,2) if isinstance(x,float) else x)).to_dict())
                value_list = list(
                    pdf(pd.concat([pd.DataFrame(self.日pe[self.日pe_c1]), pd.DataFrame(self.单笔pe[self.单笔pe_c1])]),
                        float=2)[0].apply(lambda x: round(x, 2) if isinstance(x, float) else x))
                print(value_list)
                # name_list = self.日pe_c1
                # info_list = [":".join([name_list[index], str(value_list[index])]) for index in range(0, len(name_list))]
                # print(info_list)
                # pass

        else:
            print((pdf(pd.DataFrame(self.日pe[self.日pe_c1]), 200, float=3)[0].apply(
                lambda x: round(x, 2) if isinstance(x, float) else x)).to_dict())

    def 简要pe_v1(self):
        # return instead of print
        f = 3
        pd.options.display.float_format = '{:,.3f}'.replace('3f', f'{f}f').format
        # if isinstance(self.单笔pe, (pd.DataFrame, pd.Series)):
        # if not getattr(self, '单笔pe').empty:
        # print(self.单笔pe[self.单笔pe_c1])
        # if isinstance(self.日pe, (pd.DataFrame, pd.Series)):
        # if not getattr(self, '日pe').empty:
        # print(self.日pe[self.日pe_c1])

        if isinstance(self.单笔pe, (pd.DataFrame, pd.Series)):
            if not getattr(self, '单笔pe').empty:
                # return (
                #     list(pdf(pd.concat([pd.DataFrame(self.日pe[self.日pe_c1]), pd.DataFrame(self.单笔pe[self.单笔pe_c1])]),
                #              float=2)[0].apply(lambda x: round(x, 2) if isinstance(x, float) else x)))
                # pass
                return ((pdf(pd.concat([pd.DataFrame(self.日pe[self.日pe_c2]), pd.DataFrame(self.单笔pe[self.单笔pe_c2])]),
                           float=2)[0].apply(lambda x: round(x, 2) if isinstance(x, float) else x)).to_dict())
        else:
            # return (list(pdf(pd.DataFrame(self.日pe[self.日pe_c1]), 200, float=3)[0].apply(
            #     lambda x: round(x, 2) if isinstance(x, float) else x)))
            return (pdf(pd.DataFrame(self.日pe[self.日pe_c2]), 200, float=3)[0].apply(
                 lambda x: round(x, 2) if isinstance(x, float) else x).to_dict())

    def p6(self, h=800, w=300, 简要pe='简要pe'):

        pe_func = getattr(self, 简要pe)
        pe_func()  # 这句是简单版的pe输出

        时期 = f"{self.日收益回撤df.index[0].date()} 至 {self.日收益回撤df.index[-1].date()}"

        (self.日收益回撤df[['日累计净利润', '日累计最大回撤金额']] / 10000).iplot(subplots=True, shared_xaxes=True, shared_yaxes=False,
                                                              dimensions=(h, w),
                                                              shape=(4, 1),
                                                              specs=[[{'rowspan': 3}], [None], [None], [{}], ],
                                                              subplot_titles=['总盈利', '回撤'],

                                                              title=时期, yTitle='单位:万',
                                                              colors='blue', )  # annotations={'2015-09-01':'股指开始限仓'}
        # (self.日收益回撤df.日累计最大回撤金额 / 10000).iplot(title='回撤', yTitle='单位：万', colors='blue')
        # (self.日收益回撤df[['累计净利润初始资金收益率', '日初始资金最大回撤比率']]).iplot(subplots=True, shared_xaxes=True, shared_yaxes=False,
        #                                                 # dimensions=(1500, 600),
        #                                                 shape=(4, 1),
        #                                                 specs=[[{'rowspan': 3}], [None], [None], [{}]],
        #                                                 subplot_titles=['总收益率(单利)', '相对初始资金最大回撤率'],
        #
        #                                                 title=时期, yTitle='单位:百分比',
        #                                                 colors='blue', )  # annotations={'2015-09-01':'股指开始限仓'}
        #
        df_year = self.日收益回撤df.groupby(pd.Grouper(freq="1Y")).cumsum()
        (df_year['日净利润'] / 10000).iplot(dimensions=(h, w), title='年度盈亏', yTitle='单位：万',
                                        colors='blue', )  # dimensions=(1000, 300))

        if '当日交易次数' in self.日收益回撤df.columns:
            (self.日收益回撤df.当日交易次数).iplot(kind='bar', title='日交易次数', yTitle='单位:开平各算1次', dimensions=(h, w),
                                        colors='blue', )  # annotations={'2015-09-01':'股指开始限仓'}
            # (self.日收益回撤df.当日交易次数).iplot(dimensions=(1000, 100), kind='histogram', title='日交易次数分布', yTitle='单位:开平各算1次',
            #                             colors='blue', )  # annotations={'2015-09-01':'股指开始限仓'}

        #
        # df_day_mean = self.日收益回撤df.groupby(by=self.日收益回撤df.index.day).mean()
        # (df_day_mean['日净利润'] / 10000).iplot(dimensions=(1000, 300), kind='bar', title='所有月份日平均盈亏', yTitle='单位：万',
        #                                     colors='blue')

        (self.日收益回撤df.日净利润 / 10000).iplot(dimensions=(1000, 300), kind='bar', title='日盈亏', yTitle='单位：万', colors='blue')

        # (self.日收益回撤df.日净利润 / 10000).iplot(dimensions=(1000, 300), kind='scatter', title='日盈亏', yTitle='单位：万',
        #                                   mode='markers', size=5,
        #                                   colors='blue')

        # self.日收益回撤df_week = self.日收益回撤df.resample('W').sum()
        #
        # (self.日收益回撤df_week['日净利润'] / 10000).iplot(dimensions=(1000, 300), kind='bar', title='周盈亏', yTitle='单位：万',
        #                                           colors='blue')
        #
        self.日收益回撤df_month = self.日收益回撤df.resample('M').sum()
        #
        (self.日收益回撤df_month['日净利润'] / 10000).iplot(dimensions=(h, w), kind='bar', title='月盈亏', yTitle='单位：万',
                                                   colors='blue')

        # df_month_mean = self.日收益回撤df_month.groupby(by=self.日收益回撤df_month.index.month).mean()
        # (df_month_mean['日净利润'] / 10000).iplot(dimensions=(1000, 300), kind='bar', title='所有年份月度平均盈亏', yTitle='单位：万',
        #                                       colors='blue')

        # df_year = self.日收益回撤df.resample('Y').sum()
        #
        # df_year.index = df_year.index.to_period()
        # (df_year['日净利润'] / 10000).iplot(dimensions=(600, 300), kind='bar', title='年度盈亏', yTitle='单位：万', colors='blue')
        #
        # (self.日收益回撤df.日净利润 / 10000).iplot(dimensions=(1000, 300), kind='histogram', title='日盈亏分布', yTitle='累计次数',
        #                                   xTitle='单位：万', colors='blue')
        #
        # (self.日收益回撤df_week.日净利润 / 10000).iplot(dimensions=(1000, 300), kind='histogram', title='周盈亏分布', yTitle='累计次数',
        #                                   xTitle='单位：万', colors='blue')
        #
        # (self.日收益回撤df_month.日净利润 / 10000).iplot(dimensions=(h,w), kind='histogram', title='月盈亏分布', yTitle='累计次数',
        #                                   xTitle='单位：万', colors='blue')
        return self.详细pe()


    def p7(self, h=800, w=300):

        # self.简要pe()

        # 时期 = f"{self.日收益回撤df.index[0].date()} 至 {self.日收益回撤df.index[-1].date()}"
        #
        # (self.日收益回撤df[['日累计净利润', '日累计最大回撤金额']] / 10000).iplot(subplots=True, shared_xaxes=True, shared_yaxes=False,
        #                                                       dimensions=(h,w),
        #                                                       shape=(4, 1),
        #                                                       specs=[[{'rowspan': 3}], [None], [None], [{}], ],
        #                                                       subplot_titles=['总盈利', '回撤'],
        #
        #                                                       title=时期, yTitle='单位:万',
        #                                                       colors='blue', )  # annotations={'2015-09-01':'股指开始限仓'}
        # df_year = self.日收益回撤df.groupby(pd.Grouper(freq="1Y")).cumsum()
        # (df_year['日净利润'] / 10000).iplot(dimensions=(h,w),title='年度盈亏', yTitle='单位：万', colors='blue',) #dimensions=(1000, 300))
        #
        # if '当日交易次数' in self.日收益回撤df.columns:
        #     (self.日收益回撤df.当日交易次数).iplot(kind='bar', title='日交易次数', yTitle='单位:开平各算1次', dimensions=(h,w),
        #                                 colors='blue', )  # annotations={'2015-09-01':'股指开始限仓'}
        # self.日收益回撤df_month = self.日收益回撤df.resample('M').sum()
        # #
        # (self.日收益回撤df_month['日净利润'] / 10000).iplot(dimensions=(h,w), kind='bar', title='月盈亏', yTitle='单位：万',
        #                                            colors='blue')
        for para in list(self.parameters.columns)[4:]:
            self.parameters[para].iplot(kind='box', boxpoints='all', dimensions=(h, w))

        # if type(self.单笔收益) == pd.core.series.Series:
        #     return pdf(pd.concat([pd.DataFrame(self.日pe), pd.DataFrame(self.单笔pe)]), 200,float=3), pdf(self.parameters, 200,float=3)
        # else:
        #     return pdf(pd.DataFrame(self.日pe), 200,float=3), pdf(self.parameters, 200,float=3)
        return pdf(self.parameters, 200, float=3)

    def p8(self, h=800, w=300):
        pass


    def _日收益回撤(self, 单笔收益=None, 日收益=None, 日收益按交易日补缺=True):
        日初始资金 = self.初始资金
        df_day = pd.DataFrame()

        if isinstance(单笔收益, pd.Series):
            # 日收益不能再次计算单笔滑点,用单笔算好减过滑点的净利润算日收益
            # if isinstance(单笔收益,pd.Series):
            #     if self.单次滑点单位元:
            #         单笔收益 = 单笔收益 - (self.单次滑点单位元 * 2)

            df_day_pre = 单笔收益
            df_day['当日交易次数'] = df_day_pre.groupby('日期').count() * 2
            日收益 = df_day_pre.groupby('日期').sum()

        # elif isinstance(日收益,pd.Series):
        #     df_day['日净利润']  = 日收益

        if 日收益按交易日补缺:
            # 日收益 = 日收益.resample('B').ffill()  # 按工作日重采样，向前填充这个不是A股交易日
            # 如果定义了起始日期且当起始日期大于策略起始日期时, 更改起始日期, 否则是策略起始日期
            start_date = self.startTime if self.startTime and pd.Timestamp(self.startTime) > pd.Timestamp(
                日收益.index[0]) else pd.Timestamp(日收益.index[0])
            # 如果定义了终止日期且当终止日期小于策略终止日期时, 更改终止日期, 否则是策略终止日期
            end_date = self.endTime if self.endTime and pd.Timestamp(self.endTime) < pd.Timestamp(
                日收益.index[0]) else pd.Timestamp(日收益.index[-1])

            # xnys = tc.get_calendar("XSHG")
            # XSHG_trading_calendar = xnys.sessions_in_range(start, end)
            # 日收益 = 日收益.reindex(pd.to_datetime(XSHG_trading_calendar.date), fill_value=0)
            trading_calendar = jq.get_trade_days(start_date, end_date)

            日收益 = 日收益.reindex(pd.to_datetime(trading_calendar), fill_value=0)
            # reindex 后没有 index name 了, 所以手动补全
            日收益.index.rename("日期", inplace=True)

        # df_day = df_day.reindex(pd.to_datetime(XSHG_trading_calendar.date), fill_value=0)
        df_day = df_day.reindex(pd.to_datetime(trading_calendar), fill_value=0)

        df_day['日净利润'] = 日收益
        # df_day.rename(columns={'净利润': '日净利润'}, inplace=True)
        df_day['日累计净利润'] = df_day['日净利润'].cumsum()
        df_day['日权益'] = df_day['日累计净利润'] + 日初始资金
        df_day['期初权益'] = df_day['日累计净利润'] - df_day['日净利润'] + 日初始资金
        # df_day['日累计收益率'] = df_day['日累计净利润'] * 100 / 日初始资金    #修改为  累计净利润初始资金收益率

        # df_day['日收益率'] = (df_day['日净利润'] / df_day['日权益'].shift(1))
        df_day['累计净利润初始资金收益率'] = df_day['日累计净利润'] * 100 / 日初始资金
        df_day['日收益率累加'] = (df_day['日净利润'] / df_day['期初权益']).cumsum() * 100  # df_day['日累计净利润'] * 100 / 日初始资金
        # df_day['日收益率(单利v1)'] = (df_day['日净利润'] / df_day['日权益'].shift(1))  # 单利
        df_day['日收益率(单利)'] = (df_day['日净利润'] / df_day['期初权益'])  # 单利
        df_day['日收益率(复利)'] = (df_day['日净利润'] * (df_day['期初权益'] / 日初始资金) / df_day['期初权益'])  # 复利   = df_cal['净利润']/self.初始资金
        df_day['单利净值'] = self.get_net_value(df_day['日收益率(单利)'])
        df_day['复利净值'] = self.get_net_value(df_day['日收益率(复利)'])
        df_day['日最大权益'] = df_day['日权益'].expanding(min_periods=1).max()  # 计算当日之前的账户最大价值
        df_day['日累计最大回撤金额'] = df_day['日权益'] - df_day['日最大权益']
        # df_day['日初始资金最大回撤比率'] = (df_day['日权益'] / df_day['日最大权益'] - 1) * 100  # 计算当日的回撤
        df_day['日初始资金最大回撤比率'] = df_day['日累计最大回撤金额'] / self.初始资金 * 100  # 计算当日的回撤
        df_day['日累计最大回撤比率'] = df_day['日累计最大回撤金额'] / df_day['日权益'] * 100

        self.日收益回撤df = df_day
        self.日收益回撤df.strat_name = self.strat_name
        self.日收益回撤df.strat_allname = self.strat_allname

    def 基于单笔收益计算性能(self, 单笔收益):

        # 计算年化收益率函数(全年收益)
        # 年化收益率是指投资期限为一年所获的收益率。
        # 年收益率 = [（投资内收益 / 本金） / 投资笔数] *365 ×100 %
        # 年化收益 = 本金×年化收益率
        # 实际收益 = 本金×年化收益率×投资笔数 / 365

        # def 单笔均收益乘250(dfx):
        #     单笔均收益乘250 = dfx['单笔净利润'].mean() * 250
        #     单笔均收益乘250除初始资金 = dfx['单笔净利润'].mean() * 250 / dfx['期初权益'].iloc[0]
        #     performance['单笔均收益*250'] = 单笔均收益乘250
        #     performance['单笔均收益乘250除初始资金'] = 单笔均收益乘250除初始资金

        #     print('单笔均收益乘250 ',单笔均收益乘250 )
        #     print('单笔均收益乘250除初始资金',单笔均收益乘250除初始资金)
        # 计算年化单笔累计收益率函数  计算公式为：（账户最终价值/账户初始价值）^（250/回测期间总笔数）-1
        def annual_return(date_line, capital_line):
            """
            :param date_line: 单笔期序列
            :param capital_line: 账户价值序列
            :return: 输出在回测期间的年化收益率
            """
            # 将数据序列合并成dataframe并按单笔期排序
            df_def = pd.DataFrame({'date': date_line, 'capital': capital_line})
            df_def.sort_values(by='date', inplace=True)
            df_def.reset_index(drop=True, inplace=True)
            rng = pd.period_range(df_def['date'].iloc[0], df_def['date'].iloc[-1], freq='D')
            # 计算年化收益率
            annual = pow(
                df_def.iloc[len(df_def.index) - 1, df_def.columns.get_loc('capital')] / df_def.iloc[
                    0, df_def.columns.get_loc('capital')],
                250 / len(rng)) - 1
            performance['年化收益率_vx'] = annual

        #     print('\n年化收益率：%f' % annual)
        #     print('计算公式为：（账户最终价值/账户初始价值）^（250/回测期间总笔数）-1')
        #     performance_self['逐单笔年化收益率为'] = annual

        # 计算最大回撤函数
        def max_drawdown(date_line, capital_line):
            """
            :param date_line: 单笔期序列
            :param capital_line: 账户价值序列
            :return: 输出最大回撤及开始单笔期和结束单笔期
            """
            # 将数据序列合并为一个dataframe并按单笔期排序
            df_def = pd.DataFrame({'date': date_line, 'capital': capital_line})
            df_def.sort_values(by='date', inplace=True)
            df_def.reset_index(drop=True, inplace=True)
            df_def['max2here'] = df_def['capital'].expanding(min_periods=1).max()  # 计算当单笔之前的账户最大价值
            df_def['dd2here'] = df_def['capital'] / df_def['max2here'] - 1  # 计算当单笔的回撤
            df_def['dd2hereM'] = df_def['capital'] - df_def['max2here']
            # 计算最大回撤和结束时间
            temp = df_def.sort_values(by='dd2hereM').iloc[0][['date', 'dd2here', 'dd2hereM']]
            max_dd = temp['dd2here']
            end_date = temp['date']
            max_dd_M = temp['dd2hereM']

            # 计算开始时间
            df_def = df_def[df_def['date'] <= end_date]
            start_date = df_def.sort_values(by='capital', ascending=False).iloc[0]['date']
            performance['最大回撤率为'] = max_dd
            performance['最大回撤金额为'] = max_dd_M
            performance['最大回撤率开始单笔'] = start_date
            performance['最大回撤率结束单笔'] = end_date

        #     print('\n最大回撤率为：%f, 金额:%s 开始单笔期：%s, 结束单笔期：%s' % (max_dd,max_dd_M , start_date, end_date))
        #     print('公式:min(账户当单笔价值 / 当单笔之前账户最高价值-1)')

        # 计算平均涨幅
        # def average_change(date_line, return_line):
        #     """
        #     :param date_line: 单笔期序列
        #     :param return_line: 账户单笔收益率序列
        #     :return: 输出平均涨幅
        #     """
        #     df_def = pd.DataFrame({'date': date_line, 'rtn': return_line})
        #     ave = df_def['rtn'].mean()
        #     performance['平均涨幅为'] = ave

        #     print('\n平均涨幅为：%f' % ave)
        #     print('计算公式为：账户单笔收益的平均值。')

        # 计算上涨概率
        # def prob_up(date_line, return_line):
        #     """
        #     :param date_line: 单笔期序列
        #     :param return_line: 账户单笔收益率序列
        #     :return: 输出上涨概率
        #     """
        #     df_def = pd.DataFrame({'date': date_line, 'rtn': return_line})
        #     df_def.loc[df_def[df_def['rtn'] > 0].index, 'rtn'] = 1  # 收益率大于0的记为1
        #     df_def.loc[df_def[df_def['rtn'] <= 0].index, 'rtn'] = 0  # 收益率小于等于0的记为0
        #     # 统计1和0各出现的次数
        #     count = df_def['rtn'].value_counts()
        #     p_up = count.loc[1] / len(df_def.index)
        #     performance['上涨概率为'] = p_up

        #     print('\n上涨概率为：%f' % p_up)
        #     print('计算公式为：上涨笔数 / 回测交易单笔数量。')

        # 计算最大连续上涨笔数和最大连续下跌笔数
        # def max_successive_up(date_line, return_line):
        #     """
        #     :param date_line: 单笔期序列
        #     :param return_line: 账户单笔收益率序列
        #     :return: 输出最大连续上涨笔数和最大连续下跌笔数
        #     """
        #     df_def = pd.DataFrame({'date': date_line, 'rtn': return_line})
        #     # 新建一个全为空值的series,并作为dataframe新的一列
        #     s = pd.Series(np.nan, index=df_def.index)
        #     s.name = 'up'
        #     df_def = pd.concat([df_def, s], axis=1)
        #
        #     # 当收益率大于0时，up取1，小于0时，up取0，等于0时采用前向差值
        #     df_def.loc[df_def[df_def['rtn'] > 0].index, 'up'] = 1
        #     df_def.loc[df_def[df_def['rtn'] < 0].index, 'up'] = 0
        #     df_def['up'].fillna(method='ffill', inplace=True)
        #
        #     # 根据up这一列计算到某天为止连续上涨下跌的笔数
        #     rtn_list = list(df_def['up'])
        #     successive_up_list = []
        #     num = 1
        #     for i in range(len(rtn_list)):
        #         if i == 0:
        #             successive_up_list.append(num)
        #         else:
        #             if (rtn_list[i] == rtn_list[i - 1] == 1) or (rtn_list[i] == rtn_list[i - 1] == 0):
        #                 num += 1
        #             else:
        #                 num = 1
        #             successive_up_list.append(num)
        #     # 将计算结果赋给新的一列'successive_up'
        #     df_def['successive_up'] = successive_up_list
        #     # 分别在上涨和下跌的两个dataframe里按照'successive_up'的值排序并取最大值
        #     try:
        #         max_successive_up = \
        #             df_def[df_def['up'] == 1].sort_values(by='successive_up', ascending=False)['successive_up'].iloc[0]
        #     except IndexError:
        #         max_successive_up = 1
        #     try:
        #         max_successive_down = \
        #             df_def[df_def['up'] == 0].sort_values(by='successive_up', ascending=False)['successive_up'].iloc[0]
        #     except IndexError:
        #         max_successive_down = 1
        #     # max_successive_up = \
        #     #     df_def[df_def['up'] == 1].sort_values(by='successive_up', ascending=False)['successive_up'].iloc[0]
        #     # max_successive_down = \
        #     #     df_def[df_def['up'] == 0].sort_values(by='successive_up', ascending=False)['successive_up'].iloc[0]
        #     performance['最大连续盈利笔数为'] = max_successive_up
        #     performance['最大连续亏损笔数为'] = max_successive_down

        #     print('最大连续上涨笔数为：%d  最大连续下跌笔数为：%d' % (max_successive_up, max_successive_down))

        # 计算最大单周期涨幅和最大单周期跌幅
        # def max_period_return(date_line, return_line):
        #     """
        #     :param date_line: 单笔期序列
        #     :param return_line: 账户单笔收益率序列
        #     :return: 输出最大单周期涨幅和最大单周期跌幅
        #     """
        #     df_def = pd.DataFrame({'date': date_line, 'rtn': return_line})
        #     # 分别计算单笔收益率的最大值和最小值
        #     max_return = df_def['rtn'].max()
        #     min_return = df_def['rtn'].min()
        #     performance['最大单周期涨幅为'] = max_return
        #     performance['最大单周期跌幅为'] = min_return

        #     print('最大单周期涨幅为：%f  最大单周期跌幅为：%f' % (max_return, min_return))

        # 计算收益波动率的函数
        # def volatility(date_line, return_line):
        #     """
        #     :param date_line: 单笔期序列
        #     :param return_line: 账户单笔收益率序列
        #     :return: 输出回测期间的收益波动率
        #     """
        #     from math import sqrt
        #     df_def = pd.DataFrame({'date': date_line, 'rtn': return_line})
        #     # 计算波动率
        #     vol = df_def['rtn'].std() * sqrt(250)
        #     performance['收益波动率为'] = vol

        #     print('\n收益波动率为：%f' % vol)
        #     print('收益波动率就衡量了策略收益的波动情况即风险。计算公式为：账户单笔收益的年化标准差')

        # 计算贝塔的函数
        # def beta(date_line, return_line, indexreturn_line):
        #     """
        #     :param date_line: 单笔期序列
        #     :param return_line: 账户单笔收益率序列
        #     :param indexreturn_line: 指数的收益率序列
        #     :return: 输出beta值
        #     """
        #     df_def = pd.DataFrame({'date': date_line, 'rtn': return_line, 'benchmark_rtn': indexreturn_line})
        #     # 账户收益和基准收益的协方差除以基准收益的方差
        #     b = df_def['rtn'].cov(df_def['benchmark_rtn']) / df_def['benchmark_rtn'].var()
        #     #     performance['beta'] = b
        #     print('beta: %f' % b)

        # 计算alpha的函数
        # def alpha(date_line, capital_line, index_line, return_line, indexreturn_line):
        #     """
        #     :param date_line: 单笔期序列
        #     :param capital_line: 账户价值序列
        #     :param index_line: 指数序列
        #     :param return_line: 账户单笔收益率序列
        #     :param indexreturn_line: 指数的收益率序列
        #     :return: 输出alpha值
        #     """
        #     # 将数据序列合并成dataframe并按单笔期排序
        #     df_def = pd.DataFrame(
        #         {'date': date_line, 'capital': capital_line, 'benchmark': index_line, 'rtn': return_line,
        #          'benchmark_rtn': indexreturn_line})
        #     df_def.sort_values(by='date', inplace=True)
        #     df_def.reset_index(drop=True, inplace=True)
        #     rng = pd.period_range(df_def['date'].iloc[0], df_def['date'].iloc[-1], freq='D')
        #     rf = 0.0284  # 无风险利率取10年期国债的到期年化收益率
        #
        #     annual_stock = pow(df_def.iloc[len(df_def.index) - 1, df_def.columns.get_loc('capital')] / df_def.iloc[
        #         0, df_def.columns.get_loc('capital')], 250 / len(rng)) - 1  # 账户年化收益
        #     annual_index = pow(df_def.iloc[len(df_def.index) - 1, df_def.columns.get_loc('benchmark')] / df_def.iloc[
        #         0, df_def.columns.get_loc('benchmark')], 250 / len(rng)) - 1  # 基准年化收益
        #
        #     beta = df_def['rtn'].cov(df_def['benchmark_rtn']) / df_def['benchmark_rtn'].var()  # 计算贝塔值
        #     a = (annual_stock - rf) - beta * (annual_index - rf)  # 计算alpha值
        #     performance['alpha'] = a

        #     print('alpha：%f' % a)

        # 计算夏普比函数
        def sharpe_ratio(date_line, capital_line, return_line):
            """
            :param date_line: 单笔期序列
            :param capital_line: 账户价值序列
            :param return_line: 账户单笔收益率序列
            :return: 输出夏普比率
            """
            from math import sqrt
            # 将数据序列合并为一个dataframe并按单笔期排序
            df_def = pd.DataFrame({'date': date_line, 'capital': capital_line, 'rtn': return_line})
            df_def.sort_values(by='date', inplace=True)
            df_def.reset_index(drop=True, inplace=True)
            rng = pd.period_range(df_def['date'].iloc[0], df_def['date'].iloc[-1], freq='D')
            # rf = 0.0284  # 无风险利率取10年期国债的到期年化收益率
            rf = 0
            # 账户年化收益
            annual_stock = pow(df_def.iloc[len(df_def.index) - 1, df_def.columns.get_loc('capital')] / df_def.iloc[
                0, df_def.columns.get_loc('capital')], 250 / len(rng)) - 1
            # 计算收益波动率
            volatility = df_def['rtn'].std() * sqrt(250)
            # 计算夏普比
            sharpe = (annual_stock - rf) / volatility
            performance['sharpe_ratio'] = f"{sharpe:.2f}"

        #     print('sharpe_ratio: %f' % sharpe)

        # 计算信息比率函数
        def info_ratio(date_line, return_line, indexreturn_line):
            """
            :param date_line: 单笔期序列
            :param return_line: 账户单笔收益率序列
            :param indexreturn_line: 指数的收益率序列
            :return: 输出信息比率
            """
            from math import sqrt
            df_def = pd.DataFrame({'date': date_line, 'rtn': return_line, 'benchmark_rtn': indexreturn_line})
            df_def['diff'] = df_def['rtn'] - df_def['benchmark_rtn']
            annual_mean = df_def['diff'].mean() * 250
            annual_std = df_def['diff'].std() * sqrt(250)
            info = annual_mean / annual_std
            print('info_ratio: %f' % info)

        # 计算股票和基准在回测期间的累计收益率并画图
        def cumulative_return(date_line, return_line, indexreturn_line):
            """
            :param date_line: 单笔期序列
            :param return_line: 账户单笔收益率序列
            :param indexreturn_line: 指数单笔收益率序列
            :return: 画出股票和基准在回测期间的累计收益率的折线图
            """
            df_def = pd.DataFrame({'date': date_line, 'rtn': return_line, 'benchmark_rtn': indexreturn_line})
            df_def['stock_cumret'] = (df_def['rtn'] + 1).cumprod()
            df_def['benchmark_cumret'] = (df_def['benchmark_rtn'] + 1).cumprod()
            # 画出股票和基准在回测期间的累计收益率的折线图
            df_def['stock_cumret'].plot(style='k-', figsize=(12, 5))
            df_def['benchmark_cumret'].plot(style='k--', figsize=(12, 5))
            plt.show()

        # 计算单笔数据

        df_day = pd.DataFrame()
        df_day['单笔净利润'] = 单笔收益
        df_day['单笔累计净利润'] = df_day['单笔净利润'].cumsum()
        df_day['单笔权益'] = df_day['单笔累计净利润'] + self.初始资金
        df_day['期初权益'] = df_day['单笔累计净利润'] - df_day['单笔净利润'] + self.初始资金
        # df_day['日累计收益率'] = df_day['单笔累计净利润'] / self.初始资金
        df_day['单笔收益率'] = (df_day['单笔净利润'] / df_day['单笔权益'].shift(1))
        # df_day['日最大权益'] = df_day['单笔权益'].expanding(min_periods=1).max()  # 计算当单笔之前的账户最大价值
        # df_day['单笔累计最大回撤金额'] = df_day['单笔权益'] - df_day['日最大权益']
        # df_day['单笔累计最大回撤比率'] = df_day['单笔权益'] / df_day['日最大权益'] - 1  # 计算当单笔的回撤

        performance = {}
        performance['策略名'] = self.strat_name
        performance['策略全名'] = self.strat_allname
        df_day = df_day.sort_values(by='日期')
        date_line, capital_line, return_line, = df_day.index, df_day['单笔权益'], df_day[
            '单笔收益率']  # index_line, indexreturn_line
        index_line, indexreturn_line = df_day['单笔权益'], df_day['单笔收益率']

        # 计算当前交易日日历
        # xnys = tc.get_calendar("XSHG")
        # trading_calendar = xnys.sessions_in_range(pd.Timestamp(date_line[0]), pd.Timestamp(date_line[-1]))
        start_date = pd.Timestamp(date_line[0])
        end_date = pd.Timestamp(date_line[-1])
        trading_calendar = get_trade_date(start_date, end_date)
        # print(df_day)
        # 年化收益率
        # 单笔均收益乘250(df_day)
        # annual_return(date_line, capital_line)
        # 平均涨幅
        # average_change(date_line, return_line)
        # 上涨概率
        # prob_up(date_line, return_line)
        # 最大连续上涨笔数和最大连续下跌笔数
        # max_successive_up(date_line, return_line)
        # 最大单周期涨幅和最大单周期跌幅
        # max_period_return(date_line, return_line)
        # 收益波动率
        # volatility(date_line, return_line)
        # 最大回撤
        # dd = max_drawdown(date_line, capital_line)
        # 夏普比率
        # print('\n夏普比率')
        # sharpe_ratio(date_line, capital_line, return_line)

        # performance['单笔收益率方差'] = (df_day.单笔收益率 * 100).var()
        # performance['单笔收益率标准差'] = (df_day.单笔收益率 * 100).std()
        # performance['组合策略数'] = self.组合策略数
        # performance['初始资金'] = self.初始资金
        # performance['期末资金'] = f"{self.初始资金  + df_day['单笔净利润'].sum():.0f}"
        performance['亏损单笔平均亏损'] = df_day[df_day['单笔净利润'] < 0]['单笔净利润'].mean()
        # performance['亏损单笔平均亏损率'] =performance['亏损单笔平均亏损']/performance['初始资金']
        performance['盈利单笔平均收益'] = df_day[df_day['单笔净利润'] > 0]['单笔净利润'].mean()
        # performance['盈利单笔平均收益率'] =f"{performance['盈利单笔平均收益']/performance['初始资金']*100:.2f}%"
        performance['最大单笔亏损'] = min(df_day['单笔净利润'])
        performance['最大单笔盈利'] = max(df_day['单笔净利润'])
        performance['总交易笔数'] = len(df_day)
        # todo
        performance['交易日日平均交易笔数'] = len(df_day) / len(trading_calendar)
        performance['总滑点金额'] = len(df_day) * self.单次滑点单位元 * 2
        # performance['盈利笔数'] = pd.value_counts(df_day['单笔净利润'] > 0)[True]
        # performance['亏损笔数'] = pd.value_counts(df_day['单笔净利润']<0)[True]
        performance['盈利笔数'] = sum(df_day['单笔净利润'] > 0)
        performance['亏损笔数'] = sum(df_day['单笔净利润'] < 0)
        if isinstance(self.单笔收益trade_df, pd.DataFrame) and '方向' in self.单笔收益trade_df.columns:
            performance['开多笔数'] = len(self.单笔收益trade_df.query("方向==1"))
            if performance['开多笔数']:
                performance['开多胜率'] = len(self.单笔收益trade_df.query("方向==1").query("净利润>0")) / performance['开多笔数']
                performance['开多盈亏比'] = -self.单笔收益trade_df.query("方向==1").query("净利润>0").净利润.sum() / self.单笔收益trade_df. \
                    query("方向==1").query("净利润<0").净利润.sum()
            else:
                performance['开多胜率'] = 0
                performance['开多盈亏比'] = 0

            performance['开空笔数'] = len(self.单笔收益trade_df.query("方向==2"))
            if performance['开空笔数'] != 0:
                performance['开空胜率'] = len(self.单笔收益trade_df.query("方向==2").query("净利润>0")) / performance['开空笔数']
                performance['开空盈亏比'] = -self.单笔收益trade_df.query("方向==2").query("净利润>0").净利润.sum() / self.单笔收益trade_df. \
                    query("方向==2").query("净利润<0").净利润.sum()
            else:
                performance['开空胜率'] = 0
                performance['开空盈亏比'] = 0
        # performance['大于1000盈利天'] = pd.value_counts(df_day['单笔净利润']>1000)[True]
        # performance['大于5000盈利天'] = pd.value_counts(df_day['单笔净利润']>5000)[True]
        # performance['大于10000盈利天'] = pd.value_counts(df_day['单笔净利润']>10000)[True]
        # performance['小于-1000亏损天'] = pd.value_counts(df_day['单笔净利润']<-1000)[True]
        # performance['小于-5000亏损天'] = pd.value_counts(df_day['单笔净利润']<-5000)[True]
        # performance['小于-10000亏损天'] = pd.value_counts(df_day['单笔净利润']<-10000)[True]

        performance['单笔胜率'] = f"{(df_day.shape[0]-performance['亏损笔数']) /df_day.shape[0]*100:.2f}%"
        performance['单笔盈亏比'] = f"{-performance['盈利单笔平均收益'] / performance['亏损单笔平均亏损']:.2f}%"
        performance['单笔平均收益'] = df_day['单笔净利润'].mean()

        performance['单笔平均收益率百分之'] = f"{performance['单笔平均收益']/self.初始资金*self.组合策略数*100:.4f}%"

        self.单笔pe = pd.Series(performance)
        # self.result = pd.DataFrame(performance)
        # return self.result

    def 基于日收益计算性能(self, 日收益, 日收益按交易日补缺=True):
        # 经测试, 发现这个日历包含非交易日数据, 废弃
        xnys = tc.get_calendar("XSHG")
        if 日收益按交易日补缺:
            # 日收益 = 日收益.resample('B').ffill()  # 按工作日重采样，向前填充这个不是A股交易日
            # 如果定义了起始日期且当起始日期大于策略起始日期时, 更改起始日期, 否则是策略起始日期
            start_date = self.startTime if self.startTime and pd.Timestamp(self.startTime) > pd.Timestamp(
                日收益.index[0]) else pd.Timestamp(日收益.index[0])
            # 如果定义了终止日期且当终止日期小于策略终止日期时, 更改终止日期, 否则是策略终止日期
            end_date = self.endTime if self.endTime and pd.Timestamp(self.endTime) < pd.Timestamp(
                日收益.index[0]) else pd.Timestamp(日收益.index[-1])
            # XSHG_trading_calendar = xnys.sessions_in_range(start, end)
            trading_calendar = get_trade_date(start_date, end_date)
            # reindex 后没有 index name 了, 所以手动补全
            # 日收益 = 日收益.reindex(pd.to_datetime(XSHG_trading_calendar.date), fill_value=0)
            日收益 = 日收益.reindex(pd.to_datetime(trading_calendar), fill_value=0)
            日收益.index.rename("日期", inplace=True)

        # 计算年化收益率函数(全年收益)
        # 年化收益率是指投资期限为一年所获的收益率。
        # 年收益率 = [（投资内收益 / 本金） / 投资天数] *365 ×100 %
        # 年化收益 = 本金×年化收益率
        # 实际收益 = 本金×年化收益率×投资天数 / 365
        def 日均收益乘250(dfx):
            # TODO:

            # pip install trading-alendars
            # import trading_calendars as tc

            # trading_calendar = xnys.sessions_in_range(pd.Timestamp(dfx.index[0]), pd.Timestamp(dfx.index[-1]))
            start_date = pd.Timestamp(dfx.index[0])
            end_date = pd.Timestamp(dfx.index[-1])
            trading_calendar = get_trade_date(start_date, end_date)

            日均收益乘250 = dfx['日净利润'].sum() * 250 / len(trading_calendar)
            performance['开盘交易日'] = len(trading_calendar)
            performance['策略交易日'] = len(dfx)

            # 日均收益乘250 = dfx['日净利润'].mean() * 250
            日均收益乘250除初始资金 = 日均收益乘250 / dfx['期初权益'].iloc[0]
            # performance['平均年收益(日均收益*250交易日)'] = 日均收益乘250
            performance['初始资金年化收益'] = f"{日均收益乘250除初始资金 * 100 :.2f}%"
            # performance['累计资金年化收益'] = f"{日均收益乘250除初始资金 * 100 :.2f}%"

        #     print('日均收益乘250 ',日均收益乘250 )
        #     print('日均收益乘250除初始资金',日均收益乘250除初始资金)
        # 计算年化日累计收益率函数  计算公式为：（账户最终价值/账户初始价值）^（250/回测期间总天数）-1
        def annual_return(date_line, capital_line):
            """
            :param date_line: 日期序列
            :param capital_line: 账户价值序列
            :return: 输出在回测期间的年化收益率
            """
            # 将数据序列合并成dataframe并按日期排序
            df_def = pd.DataFrame({'date': date_line, 'capital': capital_line})
            df_def.sort_values(by='date', inplace=True)
            df_def.reset_index(drop=True, inplace=True)
            rng = pd.period_range(df_def['date'].iloc[0], df_def['date'].iloc[-1], freq='D')
            # 计算年化收益率
            annual = (pow(
                df_def.iloc[len(df_def.index) - 1, df_def.columns.get_loc('capital')] / df_def.iloc[
                    0, df_def.columns.get_loc('capital')],
                250 / len(rng)) - 1) * 100
            # performance['年化收益率'] = annual

        # 计算最大回撤函数
        def max_drawdown(date_line, capital_line):
            """
            :param date_line: 日期序列
            :param capital_line: 账户价值序列
            :return: 输出最大回撤及开始日期和结束日期
            """
            # 将数据序列合并为一个dataframe并按日期排序
            df_def = pd.DataFrame({'date': date_line, 'capital': capital_line})
            df_def.sort_values(by='date', inplace=True)
            df_def.reset_index(drop=True, inplace=True)
            df_def['max2here'] = df_def['capital'].expanding(min_periods=1).max()  # 计算当日之前的账户最大价值
            df_def['dd2here'] = df_def['capital'] / df_def['max2here'] - 1  # 计算当日的回撤
            df_def['dd2hereM'] = df_def['capital'] - df_def['max2here']
            # 计算最大回撤和结束时间
            temp = df_def.sort_values(by='dd2hereM').iloc[0][['date', 'dd2here', 'dd2hereM']]
            max_dd = temp['dd2here']
            end_date = temp['date']
            max_dd_M = temp['dd2hereM']

            # 计算开始时间
            df_def_endtime = df_def[df_def['date'] <= end_date]
            start_date = df_def_endtime.sort_values(by='capital', ascending=False).iloc[0]['date']
            performance['累计最大回撤率为'] = f"{max_dd * 100 :.2f}%"
            # performance['累计最大回撤率为'] =round(max_dd,3)
            performance['最大回撤金额为'] = round(max_dd_M, 0)
            # performance['当前回撤金额为'] = round(df_def['dd2here'].iloc[-1],0)
            performance['初始资金最大回撤率为'] = f"{max_dd_M/self.初始资金 * 100 :.2f}%"
            performance['当前回撤金额为'] = round(df_def['capital'].iloc[-1] - df_def['max2here'].iloc[-1], 0)
            performance[
                '初始资金当前回撤率为'] = f"{(df_def['capital'].iloc[-1] - df_def['max2here'].iloc[-1])/self.初始资金*100:.2f}%"
            # performance['初始资金当前回撤率为'] =(df_def['capital'].iloc[-1] - df_def['max2here'].iloc[-1])/self.初始资金

            performance['最大回撤率开始日期'] = start_date.date()
            performance['最大回撤率结束日期'] = end_date.date()

        #     print('\n最大回撤率为：%f, 金额:%s 开始日期：%s, 结束日期：%s' % (max_dd,max_dd_M , start_date, end_date))
        #     print('公式:min(账户当日价值 / 当日之前账户最高价值-1)')
        # 计算平均涨幅
        def average_change(date_line, return_line):
            """
            :param date_line: 日期序列
            :param return_line: 账户日收益率序列
            :return: 输出平均涨幅
            """
            df_def = pd.DataFrame({'date': date_line, 'rtn': return_line})
            ave = df_def['rtn'].mean()
            performance['平均涨幅为'] = ave

        #     print('\n平均涨幅为：%f' % ave)
        #     print('计算公式为：账户日收益的平均值。')

        # 计算上涨概率
        def prob_up(date_line, return_line):
            """
            :param date_line: 日期序列
            :param return_line: 账户日收益率序列
            :return: 输出上涨概率
            """
            df_def = pd.DataFrame({'date': date_line, 'rtn': return_line})
            df_def.loc[df_def[df_def['rtn'] > 0].index, 'rtn'] = 1  # 收益率大于0的记为1
            df_def.loc[df_def[df_def['rtn'] <= 0].index, 'rtn'] = 0  # 收益率小于等于0的记为0
            # 统计1和0各出现的次数
            count = df_def['rtn'].value_counts()
            p_up = count.loc[1] / len(df_def.index)
            performance['上涨概率为'] = p_up

        #     print('\n上涨概率为：%f' % p_up)
        #     print('计算公式为：上涨天数 / 回测交易日数量。')

        # 计算最大连续上涨天数和最大连续下跌天数
        def max_successive_up(date_line, return_line):
            """
            :param date_line: 日期序列
            :param return_line: 账户日收益率序列
            :return: 输出最大连续上涨天数和最大连续下跌天数
            """
            df_def = pd.DataFrame({'date': date_line, 'rtn': return_line})
            # 新建一个全为空值的series,并作为dataframe新的一列
            s = pd.Series(np.nan, index=df_def.index)
            s.name = 'up'
            df_def = pd.concat([df_def, s], axis=1)

            # 当收益率大于0时，up取1，小于0时，up取0，等于0时采用前向差值
            df_def.loc[df_def[df_def['rtn'] > 0].index, 'up'] = 1
            df_def.loc[df_def[df_def['rtn'] < 0].index, 'up'] = 0
            df_def['up'].fillna(method='ffill', inplace=True)

            # 根据up这一列计算到某天为止连续上涨下跌的天数
            rtn_list = list(df_def['up'])
            successive_up_list = []
            num = 1
            for i in range(len(rtn_list)):
                if i == 0:
                    successive_up_list.append(num)
                else:
                    if (rtn_list[i] == rtn_list[i - 1] == 1) or (rtn_list[i] == rtn_list[i - 1] == 0):
                        num += 1
                    else:
                        num = 1
                    successive_up_list.append(num)
            # 将计算结果赋给新的一列'successive_up'
            df_def['successive_up'] = successive_up_list
            # 分别在上涨和下跌的两个dataframe里按照'successive_up'的值排序并取最大值
            try:
                max_successive_up = \
                    df_def[df_def['up'] == 1].sort_values(by='successive_up', ascending=False)['successive_up'].iloc[0]
            except IndexError:
                max_successive_up = 1
            try:
                max_successive_down = \
                    df_def[df_def['up'] == 0].sort_values(by='successive_up', ascending=False)['successive_up'].iloc[0]
            except IndexError:
                max_successive_down = 1

            performance['最大连续盈利天数为'] = max_successive_up
            performance['最大连续亏损天数为'] = max_successive_down

        #     print('最大连续上涨天数为：%d  最大连续下跌天数为：%d' % (max_successive_up, max_successive_down))

        # 计算最大单周期涨幅和最大单周期跌幅
        def max_period_return(date_line, return_line):
            """
            :param date_line: 日期序列
            :param return_line: 账户日收益率序列
            :return: 输出最大单周期涨幅和最大单周期跌幅
            """
            df_def = pd.DataFrame({'date': date_line, 'rtn': return_line})
            # 分别计算日收益率的最大值和最小值
            max_return = df_def['rtn'].max()
            min_return = df_def['rtn'].min()
            # performance['最大单周期涨幅为'] = max_return
            # performance['最大单周期跌幅为'] = min_return

        #     print('最大单周期涨幅为：%f  最大单周期跌幅为：%f' % (max_return, min_return))

        # 计算收益波动率的函数
        def volatility(date_line, return_line):
            """
            :param date_line: 日期序列
            :param return_line: 账户日收益率序列
            :return: 输出回测期间的收益波动率
            """
            from math import sqrt
            df_def = pd.DataFrame({'date': date_line, 'rtn': return_line})
            # 计算波动率
            vol = df_def['rtn'].std() * sqrt(250)
            performance['收益波动率为'] = vol

        #     print('\n收益波动率为：%f' % vol)
        #     print('收益波动率就衡量了策略收益的波动情况即风险。计算公式为：账户日收益的年化标准差')

        # 计算贝塔的函数
        def beta(date_line, return_line, indexreturn_line):
            """
            :param date_line: 日期序列
            :param return_line: 账户日收益率序列
            :param indexreturn_line: 指数的收益率序列
            :return: 输出beta值
            """
            df_def = pd.DataFrame({'date': date_line, 'rtn': return_line, 'benchmark_rtn': indexreturn_line})
            # 账户收益和基准收益的协方差除以基准收益的方差
            b = df_def['rtn'].cov(df_def['benchmark_rtn']) / df_def['benchmark_rtn'].var()
            #     performance['beta'] = b
            print('beta: %f' % b)

        # 计算alpha的函数
        def alpha(date_line, capital_line, index_line, return_line, indexreturn_line):
            """
            :param date_line: 日期序列
            :param capital_line: 账户价值序列
            :param index_line: 指数序列
            :param return_line: 账户日收益率序列
            :param indexreturn_line: 指数的收益率序列
            :return: 输出alpha值
            """
            # 将数据序列合并成dataframe并按日期排序
            df_def = pd.DataFrame(
                {'date': date_line, 'capital': capital_line, 'benchmark': index_line, 'rtn': return_line,
                 'benchmark_rtn': indexreturn_line})
            df_def.sort_values(by='date', inplace=True)
            df_def.reset_index(drop=True, inplace=True)
            rng = pd.period_range(df_def['date'].iloc[0], df_def['date'].iloc[-1], freq='D')
            rf = 0.0284  # 无风险利率取10年期国债的到期年化收益率

            annual_stock = pow(df_def.iloc[len(df_def.index) - 1, df_def.columns.get_loc('capital')] / df_def.iloc[
                0, df_def.columns.get_loc('capital')], 250 / len(rng)) - 1  # 账户年化收益
            annual_index = pow(df_def.iloc[len(df_def.index) - 1, df_def.columns.get_loc('benchmark')] / df_def.iloc[
                0, df_def.columns.get_loc('benchmark')], 250 / len(rng)) - 1  # 基准年化收益

            beta = df_def['rtn'].cov(df_def['benchmark_rtn']) / df_def['benchmark_rtn'].var()  # 计算贝塔值
            a = (annual_stock - rf) - beta * (annual_index - rf)  # 计算alpha值
            performance['alpha'] = a

        #     print('alpha：%f' % a)

        # 计算夏普比函数
        def sharpe_ratio_v0(date_line, capital_line, return_line):
            """
            :param date_line: 日期序列
            :param capital_line: 账户价值序列
            :param return_line: 账户日收益率序列
            :return: 输出夏普比率
            """
            from math import sqrt
            # 将数据序列合并为一个dataframe并按日期排序
            df_def = pd.DataFrame({'date': date_line, 'capital': capital_line, 'rtn': return_line})
            df_def.sort_values(by='date', inplace=True)
            df_def.reset_index(drop=True, inplace=True)
            # TODO:

            # rng = xnys.sessions_in_range(pd.Timestamp(df_def['date'].iloc[0]), pd.Timestamp(df_def['date'].iloc[-1]))

            rng = pd.period_range(df_def['date'].iloc[0], df_def['date'].iloc[-1], freq='D')
            rf = 0.3  # 无风险利率取10年期国债的到期年化收益率
            # rf = 0
            # 账户年化收益
            annual_stock = pow(df_def.iloc[len(df_def.index) - 1, df_def.columns.get_loc('capital')] / df_def.iloc[
                0, df_def.columns.get_loc('capital')], 250 / len(rng)) - 1
            # print(annual_stock)
            # annual_stock = float(performance['初始资金年化收益'].strip('%')) / 100
            # 计算收益波动率
            volatility = df_def['rtn'].std() * sqrt(250)
            # performance['volatility_v0'] = f"{volatility:.2f}"
            # 计算夏普比
            sharpe = (annual_stock - rf) / volatility
            performance['sharpe_ratio_v0'] = f"{sharpe:.2f}"

        def sharpe_ratio_v1(date_line, capital_line, return_line):
            """
            :param date_line: 日期序列
            :param capital_line: 账户价值序列
            :param return_line: 账户日收益率序列
            :return: 输出夏普比率
            """
            from math import sqrt
            # 将数据序列合并为一个dataframe并按日期排序
            df_def = pd.DataFrame({'date': date_line, 'capital': capital_line, 'rtn': return_line})
            df_def.sort_values(by='date', inplace=True)
            df_def.reset_index(drop=True, inplace=True)
            # TODO:

            # rng_real = pd.DataFrame(xnys.sessions_in_range(pd.Timestamp(df_def['date'].iloc[0]), pd.Timestamp(df_def['date'].iloc[-1])))
            start_date = pd.Timestamp(df_def['date'].iloc[0])
            end_date = pd.Timestamp(df_def['date'].iloc[-1])
            trading_calendar = get_trade_date(start_date, end_date)
            rng_real = pd.DataFrame(trading_calendar)
            rng_real.columns = ['date']
            rng_real['date'] = pd.to_datetime(rng_real['date']).dt.tz_localize(None)
            rng = pd.period_range(df_def['date'].iloc[0], df_def['date'].iloc[-1], freq='D')
            # rf = 0.3  # 无风险利率取10年期国债的到期年化收益率
            rf = 0
            # 账户年化收益
            # annual_stock = pow(df_def.iloc[len(df_def.index) - 1, df_def.columns.get_loc('capital')] / df_def.iloc[
            #     0, df_def.columns.get_loc('capital')], 250 / len(rng)) - 1
            # annual_stock = float(performance['初始资金年化收益'].strip('%')) / 100
            # 计算收益波动率
            df_l = pd.merge(df_def, rng_real, on='date', how='right')
            df_l = df_l.fillna(0)
            volatility = df_l['rtn'].std() * sqrt(250)
            # performance['volatility_v1'] = f"{volatility:.2f}"
            # 计算夏普比
            sharpe = (float(performance['初始资金年化收益'].split('%')[0]) / 100 - rf) / volatility
            performance['sharpe_ratio_v1'] = f"{sharpe:.2f}"

        #     print('sharpe_ratio: %f' % sharpe)

        def sharpe_ratio_v2(df1):
            """
            date_line, capital_line, return_line
            :param date_line: 日期序列
            :param capital_line: 账户价值序列
            :param return_line: 账户日收益率序列
            :return: 输出夏普比率
            """
            from math import sqrt
            y_l = list(set(df1.index.year))
            y_l.sort()
            for i in range(0, len(set(df1.index.year))):
                dfx = df1[df1.index.year == y_l[i]]
                date_line, capital_line, return_line, = dfx.index, dfx['日权益'], dfx['日收益率']
                df_def = pd.DataFrame({'date': date_line, 'capital': capital_line, 'rtn': return_line})
                df_def['date'] = pd.to_datetime(df_def['date']).apply(lambda x: x.replace(tzinfo=None))
                df_def.sort_values(by='date', inplace=True)
                df_def.reset_index(drop=True, inplace=True)
                import pytz
                if i == 0:
                    # rng_real = pd.DataFrame(
                    #     xnys.sessions_in_range(
                    #         pd.Timestamp(df_def['date'].iloc[0]).replace(tzinfo=pytz.timezone('UTC')),
                    #         pd.Timestamp(''.join([str(df_def['date'].iloc[0].year), '-12-31'])).replace(
                    #             tzinfo=pytz.timezone('UTC'))))
                    start_date = pd.Timestamp(df_def['date'].iloc[0]).replace(tzinfo=pytz.timezone('UTC'))
                    end_date = pd.Timestamp(''.join([str(df_def['date'].iloc[0].year), '-12-31'])).replace(tzinfo=pytz.timezone('UTC'))
                    trading_calendar = get_trade_date(start_date, end_date)
                    rng_real = pd.DataFrame(trading_calendar)


                elif i == len(set(df1.index.year)) - 1:

                    # rng_real = pd.DataFrame(
                    #     xnys.sessions_in_range(
                    #         pd.Timestamp(''.join([str(df_def['date'].iloc[0].year), '-01-02'])).replace(
                    #             tzinfo=pytz.timezone('UTC')),
                    #         pd.Timestamp(df_def['date'].iloc[-1]).replace(tzinfo=pytz.timezone('UTC'))))

                    start_date = pd.Timestamp(''.join([str(df_def['date'].iloc[0].year), '-01-02'])).replace(tzinfo=pytz.timezone('UTC'))
                    end_date = pd.Timestamp(df_def['date'].iloc[-1]).replace(tzinfo=pytz.timezone('UTC'))
                    trading_calendar = get_trade_date(start_date, end_date)
                    rng_real = pd.DataFrame(trading_calendar)
                else:
                    # rng_real = pd.DataFrame(
                    #     xnys.sessions_in_range(
                    #         pd.Timestamp(''.join([str(df_def['date'].iloc[0].year), '-01-01'])).replace(
                    #             tzinfo=pytz.timezone('UTC')),
                    #         pd.Timestamp(''.join([str(df_def['date'].iloc[0].year), '-12-31'])).replace(
                    #             tzinfo=pytz.timezone('UTC'))))

                    start_date = pd.Timestamp(''.join([str(df_def['date'].iloc[0].year), '-01-01'])).replace(tzinfo=pytz.timezone('UTC'))
                    end_date = pd.Timestamp(''.join([str(df_def['date'].iloc[0].year), '-12-31'])).replace(tzinfo=pytz.timezone('UTC'))
                    trading_calendar = get_trade_date(start_date, end_date)
                    rng_real = pd.DataFrame(trading_calendar)
                # rng_real = pd.DataFrame(
                #     xnys.sessions_in_range(pd.Timestamp(df_def['date'].iloc[0]), pd.Timestamp(df_def['date'].iloc[-1])))
                # 年度日均收益乘250除初始资金=dfx['日净利润'].sum() * 250 / len(rng_real)/ dfx['期初权益'].iloc[0]
                年度日均收益乘250除初始资金 = (dfx['日收益率'] * len(dfx['日收益率']) / len(rng_real)).mean() * 250
                performance[y_l[i], '年化'] = f"{年度日均收益乘250除初始资金 * 100 :.2f}%"
                rng_real.columns = ['date']
                rng_real['date'] = pd.to_datetime(rng_real['date']).dt.tz_localize(None)

                # rng = pd.period_range(df_def['date'].iloc[0], df_def['date'].iloc[-1], freq='D')
                # rf = 0.3  # 无风险利率取10年期国债的到期年化收益率
                rf = 0
                # 账户年化收益
                # annual_stock = pow(df_def.iloc[len(df_def.index) - 1, df_def.columns.get_loc('capital')] / df_def.iloc[
                #     0, df_def.columns.get_loc('capital')], 250 / len(rng)) - 1
                # annual_stock = float(performance['初始资l金年化收益'].strip('%')) / 100
                # 计算收益波动率
                df_l = pd.merge(df_def, rng_real, on='date', how='right')
                df_l = df_l.fillna(0)
                volatility = df_l['rtn'].std() * sqrt(250)

                # 计算夏普比
                sharpe = (年度日均收益乘250除初始资金 - rf) / volatility
                performance['sharpe_ratio', y_l[i]] = f"{sharpe:.2f}"

        def sharpe_ratio_v2_del(df1):
            """
            date_line, capital_line, return_line
            :param date_line: 日期序列
            :param capital_line: 账户价值序列
            :param return_line: 账户日收益率序列
            :return: 输出夏普比率
            """
            from math import sqrt
            y_l = list(set(df1.index.year))
            y_l.sort()
            for i in range(0, len(set(df1.index.year))):
                dfx = df1[df1.index.year == y_l[i]]
                date_line, capital_line, return_line, = dfx.index, dfx['日权益'], dfx['日收益率']
                temp = pd.DataFrame({'date': date_line, 'capital': capital_line, 'rtn': return_line})
                temp.sort_values(by='date', inplace=True)
                temp.reset_index(drop=True, inplace=True)
                # print(return_line)
                temp['asset'] = (temp['rtn']).cumsum() + self.初始资金
                temp['asset_yesterday'] = [self.初始资金] + list(temp['asset'][1:])
                temp['return'] = temp['rtn'] / temp['asset_yesterday']
                arr = temp['return'].mean() * 250
                std = temp['return'].std() * np.sqrt(250)
                sharpe = arr / std
                print(arr, std, sharpe)
                performance[y_l[i], '年化'] = f"{arr * 100 :.2f}%"
                performance['sharpe_ratio', y_l[i]] = f"{sharpe:.2f}"

        def sharpe_ratio_v3(date_line, capital_line, return_line):
            """
            :param date_line: 日期序列
            :param capital_line: 账户价值序列
            :param return_line: 账户日收益率序列
            :return: 输出夏普比率
            def com_sharpe(daily_pnl):
                temp = pd.DataFrame(daily_pnl.groupby('date').pnl.sum())
                temp['asset'] = (temp['pnl']).cumsum()+1000000
                temp['asset_yesterday'] =  [1000000] + list(temp['asset'][1:])
                temp['return'] = temp['pnl']/temp['asset_yesterday']
                arr = temp['return'].mean()*250
                std = temp['return'].std()*np.sqrt(250)
                sharpe = arr/std
                return arr,sharpe
            """
            from math import sqrt
            # 将数据序列合并为一个dataframe并按日期排序
            temp = pd.DataFrame({'date': date_line, 'capital': capital_line, 'rtn': return_line})
            temp.sort_values(by='date', inplace=True)
            temp.reset_index(drop=True, inplace=True)
            # print(return_line)
            temp['asset'] = (temp['rtn']).cumsum() + self.初始资金
            temp['asset_yesterday'] = [self.初始资金] + list(temp['asset'][1:])
            temp['return'] = temp['rtn'] / temp['asset_yesterday']
            arr = temp['return'].mean() * 250
            std = temp['return'].std() * np.sqrt(250)
            sharpe = arr / std
            # print(arr,std,sharpe)
            performance['sharpe_ratio_v3'] = f"{sharpe:.4f}"
            print('sharpe_ratio_v3:', sharpe)

            # 日收益率   100日.mean()                日均收益率*250
            # 单笔       10笔.mean()     单笔收益率

        def sharpe_ratio_v4_(date_line, capital_line, return_line):
            from math import sqrt

            temp = pd.DataFrame({'date': date_line, 'capital': capital_line, 'rtn': return_line})
            temp.sort_values(by='date', inplace=True)
            temp.reset_index(drop=True, inplace=True)
            temp['rtn'].iloc[0] = (temp['capital'].iloc[0] - self.初始资金) / self.初始资金
            # print(return_line)
            temp['rtn'] = temp['rtn'] * self.初始资金
            temp['asset'] = (temp['rtn']).cumsum() + self.初始资金
            temp['asset_yesterday'] = [self.初始资金] + list(temp['asset'][1:])
            temp['return'] = temp['rtn'] / temp['asset_yesterday']
            # rng_real = pd.DataFrame(
            #     xnys.sessions_in_range(pd.Timestamp(temp['date'].iloc[0]), pd.Timestamp(temp['date'].iloc[-1])))

            start_date = pd.Timestamp(temp['date'].iloc[0])
            end_date = pd.Timestamp(temp['date'].iloc[-1])
            trading_calendar = get_trade_date(start_date, end_date)
            rng_real = pd.DataFrame(trading_calendar)

            arr = (temp['return'] * len(temp['return']) / len(rng_real)).mean() * 250
            list_0 = [0 for i in range(len(rng_real) - len(temp['return']))]
            std = pd.Series(list(temp['return']) + list_0).std() * np.sqrt(250)
            # std = temp['return'].std() * np.sqrt(250)
            try:
                sharpe = arr / std
            except:
                sharpe = 0
            # print(arr,std,sharpe)
            performance['sharpe_ratio_v5'] = f"{sharpe:.3f}"

            # print('sharpe_ratio_v4:',sharpe)

            # 日收益率   100日.mean()                日均收益率*250
            # 单笔       10笔.mean()     单笔收益率

        def sharpe_ratio_v4(df1):
            temp = df1[['期初权益', '日净利润']]
            temp = temp.reset_index()
            temp.sort_values(by='日期', inplace=True)
            temp['return'] = temp['日净利润'] / temp['期初权益']
            # rng_real = pd.DataFrame(
            #     xnys.sessions_in_range(pd.Timestamp(temp['日期'].iloc[0]), pd.Timestamp(temp['日期'].iloc[-1])))
            start_date = pd.Timestamp(temp['日期'].iloc[0])
            end_date = pd.Timestamp(temp['日期'].iloc[-1])
            trading_calendar = get_trade_date(start_date, end_date)
            rng_real = pd.DataFrame(trading_calendar)
            arr = (temp['return'] * len(temp['return']) / len(rng_real)).mean() * 250
            list_0 = [0 for i in range(len(rng_real) - len(temp['return']))]
            std = pd.Series(list(temp['return']) + list_0).std() * np.sqrt(250)
            sharpe = arr / std
            performance['sharpe_ratio_v4'] = f"{sharpe:.3f}"

        def sharpe_ratio_v5(df1):
            """
            date_line, capital_line, return_line
            :param date_line: 日期序列
            :param capital_line: 账户价值序列
            :param return_line: 账户日收益率序列
            :return: 输出夏普比率
            """
            import pytz
            y_l = list(set(df1.index.year))
            y_l.sort()
            for i in range(0, len(set(df1.index.year))):
                dfx = df1[df1.index.year == y_l[i]]
                temp = dfx[['期初权益', '日净利润']]
                if i != 0:
                    temp['期初权益'] = temp['期初权益'] - (temp['期初权益'].iloc[0] - self.初始资金)
                temp.reset_index(inplace=True)
                temp['return'] = temp['日净利润'] / temp['期初权益']
                if i == 0:
                    # rng_real = pd.DataFrame(
                    #     xnys.sessions_in_range(pd.Timestamp(temp['日期'].iloc[0]).replace(tzinfo=pytz.timezone('UTC')),
                    #                            pd.Timestamp(''.join([str(temp['日期'].iloc[0].year), '-12-31'])).replace(tzinfo=pytz.timezone('UTC'))))
                    start_date = pd.Timestamp(temp['日期'].iloc[0]).replace(tzinfo=pytz.timezone('UTC'))
                    end_date = pd.Timestamp(''.join([str(temp['日期'].iloc[0].year), '-12-31'])).replace(
                        tzinfo=pytz.timezone('UTC'))
                    trading_calendar = get_trade_date(start_date, end_date)
                    rng_real = pd.DataFrame(trading_calendar)
                elif i == len(set(df1.index.year)) - 1:

                    # rng_real = pd.DataFrame(
                    #     xnys.sessions_in_range(pd.Timestamp(''.join([str(temp['日期'].iloc[0].year), '-01-02'])).replace(tzinfo=pytz.timezone('UTC')),
                    #                            pd.Timestamp(temp['日期'].iloc[-1]).replace(tzinfo=pytz.timezone('UTC'))))
                    start_date = pd.Timestamp(''.join([str(temp['日期'].iloc[0].year), '-01-02'])).replace(
                        tzinfo=pytz.timezone('UTC'))
                    end_date = pd.Timestamp(temp['日期'].iloc[-1]).replace(tzinfo=pytz.timezone('UTC'))
                    trading_calendar = get_trade_date(start_date, end_date)
                    rng_real = pd.DataFrame(trading_calendar)
                else:
                    # rng_real = pd.DataFrame(
                    #     xnys.sessions_in_range(pd.Timestamp(''.join([str(temp['日期'].iloc[0].year), '-01-01'])).replace(tzinfo=pytz.timezone('UTC')),
                    #                            pd.Timestamp(''.join([str(temp['日期'].iloc[0].year), '-12-31'])).replace(tzinfo=pytz.timezone('UTC'))))
                    #
                    start_date = pd.Timestamp(''.join([str(temp['日期'].iloc[0].year), '-01-01'])).replace(
                        tzinfo=pytz.timezone('UTC'))
                    end_date = pd.Timestamp(''.join([str(temp['日期'].iloc[0].year), '-12-31'])).replace(
                        tzinfo=pytz.timezone('UTC'))
                    trading_calendar = get_trade_date(start_date, end_date)
                    rng_real = pd.DataFrame(trading_calendar)
                arr = (temp['return'] * len(temp['return']) / len(rng_real)).mean() * 250
                list_0 = [0 for i in range(len(rng_real) - len(temp['return']))]
                std = pd.Series(list(temp['return']) + list_0).std() * np.sqrt(250)
                if std != 0:
                    sharpe = arr / std
                else:
                    sharpe = 0
                performance[y_l[i], '年化'] = f"{arr * 100 :.2f}%"
                performance['sharpe_ratio', y_l[i]] = f"{sharpe:.2f}"

        # 计算卡尔玛比率函数
        # todo
        def Calmar_ratio(date_line, capital_line, return_line):
            """
             :param date_line: 日期序列
             :param capital_line: 账户价值序列
             :param return_line: 账户日收益率序列
             :return: 输出calmar比率
                     """
            # 将数据序列合并为一个dataframe并按日期排序
            df_def = pd.DataFrame({'date': date_line, 'capital': capital_line, 'rtn': return_line})
            df_def.sort_values(by='date', inplace=True)
            df_def.reset_index(drop=True, inplace=True)
            rng = xnys.sessions_in_range(pd.Timestamp(df_def['date'].iloc[0]), pd.Timestamp(df_def['date'].iloc[-1]))
            rng = pd.period_range(df_def['date'].iloc[0], df_def['date'].iloc[-1], freq='D')
            # rf = 0.0284  # 无风险利率取10年期国债的到期年化收益率
            rf = 0
            # 账户年化收益
            annual_stock = pow(df_def.iloc[len(df_def.index) - 1, df_def.columns.get_loc('capital')] / df_def.iloc[
                0, df_def.columns.get_loc('capital')], 250 / len(rng)) - 1

            # annual_stock = float(performance['初始资金年化收益'].strip('%'))/100
            # 计算最大回撤
            df_def['max2here'] = df_def['capital'].expanding(min_periods=1).max()  # 计算当日之前的账户最大价值
            df_def['dd2hereM'] = df_def['capital'] - df_def['max2here']
            max_drawdown_ = abs(df_def['dd2hereM'].min() / df_def.iloc[
                0, df_def.columns.get_loc('capital')])
            # 计算卡尔玛比
            # max_drawdown_ = performance['最大回撤金额为']
            calmar_ratio = (annual_stock - rf) / max_drawdown_
            performance['calmar_ratio'] = calmar_ratio

        # 计算信息比率函数
        def info_ratio(date_line, return_line, indexreturn_line):
            """
            :param date_line: 日期序列
            :param return_line: 账户日收益率序列
            :param indexreturn_line: 指数的收益率序列
            :return: 输出信息比率
            """
            from math import sqrt
            df_def = pd.DataFrame({'date': date_line, 'rtn': return_line, 'benchmark_rtn': indexreturn_line})
            df_def['diff'] = df_def['rtn'] - df_def['benchmark_rtn']
            annual_mean = df_def['diff'].mean() * 250
            annual_std = df_def['diff'].std() * sqrt(250)
            info = annual_mean / annual_std
            print('info_ratio: %f' % info)

        # 计算股票和基准在回测期间的累计收益率并画图
        def cumulative_return(date_line, return_line, indexreturn_line):
            """
            :param date_line: 日期序列
            :param return_line: 账户日收益率序列
            :param indexreturn_line: 指数日收益率序列
            :return: 画出股票和基准在回测期间的累计收益率的折线图
            """
            df_def = pd.DataFrame({'date': date_line, 'rtn': return_line, 'benchmark_rtn': indexreturn_line})
            df_def['stock_cumret'] = (df_def['rtn'] + 1).cumprod()
            df_def['benchmark_cumret'] = (df_def['benchmark_rtn'] + 1).cumprod()
            # 画出股票和基准在回测期间的累计收益率的折线图
            df_def['stock_cumret'].plot(style='k-', figsize=(12, 5))
            df_def['benchmark_cumret'].plot(style='k--', figsize=(12, 5))
            plt.show()

        # 计算日数据
        df_day = pd.DataFrame()
        df_day['日净利润'] = 日收益
        df_day['日累计净利润'] = df_day['日净利润'].cumsum()
        df_day['日权益'] = df_day['日累计净利润'] + self.初始资金
        df_day['期初权益'] = df_day['日累计净利润'] - df_day['日净利润'] + self.初始资金
        # df_day['日累计收益率'] = df_day['日累计净利润'] / self.初始资金
        df_day['日收益率'] = (df_day['日净利润'] / df_day['日权益'].shift(1))
        # df_day['日最大权益'] = df_day['日权益'].expanding(min_periods=1).max()  # 计算当日之前的账户最大价值
        # df_day['日累计最大回撤金额'] = df_day['日权益'] - df_day['日最大权益']
        # df_day['日初始资金最大回撤比率'] = df_day['日权益'] / df_day['日最大权益'] - 1  # 计算当日的回撤

        performance = {}
        performance['策略名'] = self.strat_name
        performance['策略全名'] = self.strat_allname
        # print(df_day)
        date_line, capital_line, return_line, = df_day.index, df_day['日权益'], df_day[
            '日收益率']  # index_line, indexreturn_line
        index_line, indexreturn_line = df_day['日权益'], df_day['日收益率']

        performance['起始时间'] = df_day.index[0].date()
        performance['结束时间'] = df_day.index[-1].date()
        performance['初始资金'] = self.初始资金
        performance['期末资金'] = f"{self.初始资金 + df_day['日净利润'].sum():.0f}"
        performance['总盈亏'] = float(performance['期末资金']) - float(performance['初始资金'])

        # 年化收益率
        日均收益乘250(df_day)
        annual_return(date_line, capital_line)
        # 平均涨幅
        # average_change(date_line, return_line)
        # 上涨概率
        # prob_up(date_line, return_line)
        # 最大连续上涨天数和最大连续下跌天数
        max_successive_up(date_line, return_line)
        # 最大单周期涨幅和最大单周期跌幅
        max_period_return(date_line, return_line)
        # 收益波动率
        # volatility(date_line, return_line)
        # 最大回撤
        max_drawdown(date_line, capital_line)
        # 夏普比率
        # print('\n夏普比率')
        # sharpe_ratio_v0(date_line, capital_line, return_line)
        # sharpe_ratio_v1(date_line, capital_line, return_line)
        # sharpe_ratio_v4_(date_line, capital_line, return_line)
        sharpe_ratio_v4(df_day)
        # 这个函数有0会报错
        sharpe_ratio_v5(df_day)
        Calmar_ratio(date_line, capital_line, return_line)

        # performance['日收益率方差'] = (df_day.日收益率 * 100).var()
        # performance['日收益率标准差'] = (df_day.日收益率 * 100).std()
        # performance['组合策略数'] = self.组合策略数

        # performance['盈利天数'] = pd.value_counts(df_day['日净利润'] > 0)[True]
        # performance['亏损天数'] = pd.value_counts(df_day['日净利润']<0)[True]
        performance['盈利天数'] = sum(df_day['日净利润'] > 0)
        performance['亏损天数'] = sum(df_day['日净利润'] < 0)
        # performance['大于1000盈利天'] = pd.value_counts(df_day['日净利润']>1000)[True]
        # performance['大于5000盈利天'] = pd.value_counts(df_day['日净利润']>5000)[True]
        # performance['大于10000盈利天'] = pd.value_counts(df_day['日净利润']>10000)[True]
        # performance['小于-1000亏损天'] = pd.value_counts(df_day['日净利润']<-1000)[True]
        # performance['小于-5000亏损天'] = pd.value_counts(df_day['日净利润']<-5000)[True]
        # performance['小于-10000亏损天'] = pd.value_counts(df_day['日净利润']<-10000)[True]
        performance['日胜率'] = f"{(df_day.shape[0]-performance['亏损天数']) /df_day.shape[0]*100:.2f}%"
        df_week = df_day.resample('W').sum()
        performance['周胜率'] = f"{sum(df_week['日净利润'] > 0)/ df_week.shape[0]*100:.2f}%"
        df_month = df_day.resample('M').sum()
        performance['月胜率'] = f"{sum(df_month['日净利润'] > 0)/ df_month.shape[0]*100:.2f}%"
        performance['日胜率'] = f"{(df_day.shape[0]-performance['亏损天数']) /df_day.shape[0]*100:.2f}%"
        performance['最大日亏损'] = f"{min(df_day['日净利润']):.0f}"
        performance['最大日盈利'] = f"{max(df_day['日净利润']):.0f}"

        # 所有交易日的平均收益:
        # performance['日平均收益'] =df_day['日净利润'].sum()/len(xnys.sessions_in_range(pd.Timestamp(df_day.index[0].date()), pd.Timestamp(df_day.index[-1].date()))
        # 有交易的日平均收益:
        performance['日平均收益'] = df_day['日净利润'].mean()
        performance['日平均收益率百分之'] = f"{performance['日平均收益']/self.初始资金*self.组合策略数*100:.2f}%"
        performance['亏损日平均亏损'] = df_day[df_day['日净利润'] < 0]['日净利润'].mean()
        # performance['亏损日平均亏损率'] =performance['亏损日平均亏损']/performance['初始资金']
        performance['盈利日平均收益'] = df_day[df_day['日净利润'] > 0]['日净利润'].mean()
        # performance['盈利日平均收益率'] =performance['盈利日平均收益']/performance['初始资金']
        performance['日盈亏比'] = f"{-performance['盈利日平均收益']/performance['亏损日平均亏损']:.2f}"

        self.日pe = pd.Series(performance)
        # self.result = pd.DataFrame(performance)
        # return self.result

    def disrupt_day_income(self, 日收益):
        '''随机打乱日收益率顺序'''

        def __init__(self):
            pass


class Performances():
    def reset(self):
        self.all_pe_list = self.orgin_data.copy()
        self.filtered = self.orgin_data.copy()

    def from_df_list(self, tradelist: list, date_from='2016-03-01', date_to='2050-01-01', 初始资金=1000000, 单次滑点单位元=0,
                     *kwargs):
        '''和filtered_pe的单次滑点单位元参数单独生效,合并交易后的滑点是单独用交易次数计算的,所以互不相干,不会叠加生效'''
        self.初始资金 = 初始资金
        self.单次滑点单位元 = 单次滑点单位元
        self.all_pe_list = [Performance(v, strat_name=v.strat_name, strat_allname=v.strat_allname, startTime=date_from,
                                        endTime=date_to, 初始资金=初始资金, 单次滑点单位元=self.单次滑点单位元, tradelist=v) for v in
                            tradelist]
        self.filtered = self.all_pe_list.copy()
        self.orgin_data = self.all_pe_list.copy()
        # self.tradedf = pd.concat([v.tradedf for v in self.all_pe_list])
        trade_df_list = []
        for v in self.all_pe_list:
            v.tradedf['name'] = v.strat_name
            trade_df_list.append(v.tradedf)
        self.tradedf = pd.concat(trade_df_list)
        self.tradedf.sort_index(inplace=True)
        return self

    def add_from_df_list(self, tradelist: list, date_from='2016-03-01', date_to='2050-01-01', 初始资金=1000000, 单次滑点单位元=0,
                         *kwargs):
        self.初始资金 = 初始资金
        self.单次滑点单位元 = 单次滑点单位元
        self.filtered.extend(
            [Performance(v, strat_name=v.strat_name, strat_allname=v.strat_allname, startTime=date_from,
                         endTime=date_to, 初始资金=初始资金, 单次滑点单位元=self.单次滑点单位元, tradelist=v) for v in tradelist])
        # self.all_pe_list.extend([Performance(v, strat_name=v.strat_name, strat_allname=v.strat_allname, startTime=date_from,
        #                                 endTime=date_to, 初始资金=初始资金,单次滑点单位元=self.单次滑点单位元, tradelist=v) for v in tradelist])
        # self.filtered = self.all_pe_list.copy()
        # self.orgin_data = self.all_pe_list.copy()
        return self

    def from_config(self):
        pass

    def all_name(self, nametype='strat_name'):
        '''
        nametype='strat_name'
        nametype='strat_allname'
        '''
        return [getattr(strat, nametype) for strat in self.all_pe_list]

    def concat(self, nametype='strat_name', base='filtered'):
        '''
          nametype='strat_name'
          nametype='strat_allname'

          base='all_pe_list'
          base='filtered'
          '''
        l = getattr(self, base)
        return [getattr(strat, nametype) for strat in l]

    def filter_by_pnl(self, pnl=30, base='filtered'):
        """
        base='all_pe_list'
        base='filtered'
        """
        l = getattr(self, base)
        self.filtered = [pe for pe in l if float(pe.日pe.初始资金年化收益.split('%')[0]) > pnl]
        self.del_strat = self.mddm(listname=[pe for pe in l if float(pe.日pe.初始资金年化收益.split('%')[0]) <= pnl])
        print(f'年化收益率大于{pnl}筛选前策略数量:{len(l)}, 筛选后策略数量:{len(self.filtered)}')
        # print('删掉的策略:\n')
        # return self.list_pe_to_df(attribute='日pe', listname=self.del_strat)

    def filter_by_mdd(self, mdd=30, base='filtered'):
        """
        base='all_pe_list'
        base='filtered'
        """
        l = getattr(self, base)
        self.filtered = [pe for pe in l if float(pe.日pe.初始资金最大回撤率为.split('%')[0]) > mdd]
        self.del_strat = self.mddm(listname=[pe for pe in l if float(pe.日pe.初始资金最大回撤率为.split('%')[0]) <= mdd])
        print(f'最大回撤大于{mdd}筛选前策略数量:{len(l)}, 筛选后策略数量:{len(self.filtered)}')
        # print('删掉的策略:\n')
        # return self.list_pe_to_df(attribute='日pe', listname=self.del_strat)

    def filter_by_intraday_tradetimes(self, times=10, base='filtered'):
        """
        base='all_pe_list'
        base='filtered'
        """
        l = getattr(self, base)
        self.filtered = [pe for pe in l if float(pe.单笔pe.交易日日平均交易笔数) < times]
        self.del_strat = self.mddm(listname=[pe for pe in l if float(pe.单笔pe.交易日日平均交易笔数) >= times])
        print(f'交易日日平均交易笔数小于{times}筛选前策略数量:{len(l)}, 筛选后策略数量:{len(self.filtered)}')
        # print('删掉的策略:\n')
        # return self.list_pe_to_df(attribute='日pe', listname=self.del_strat)

    def select_strat_by_name_1(self, name):
        return [n for n in self.all_pe_list if name in n.strat_name][0]

    def filter_strat_by_name(self, name, base='filtered'):
        l = getattr(self, base)
        self.filtered = [n for n in l if name in n.strat_name]
        self.del_strat = self.mddm(listname=[n for n in l if name not in n.strat_name])
        print(f'名字{name}筛选前策略数量:{len(l)}, 筛选后策略数量:{len(self.filtered)}')
        # print('删掉的策略:\n')

    def add_strat_by_name(self, name, base='all_pe_list'):
        l = getattr(self, base)
        if isinstance(name, list):
            _filtered = []
            for na in name:
                _ = [n for n in l if na in n.strat_name]
                _filtered.extend(_)

                if _ == []:
                    _ = [n for n in l if na in n.strat_allname]
                    _filtered.extend(_)
        else:
            _filtered = [n for n in l if name in n.strat_name]
            if _filtered == []:
                _filtered = [n for n in l if name in n.strat_allname]

        self.filtered.extend(_filtered)
        print(f'名字{name}添加前策略数量:{len(self.filtered) - len(_filtered)}, 添加后策略数量:{len(self.filtered)}')
        # print('删掉的策略:\n')

    def add_strat_by_pe(self, pe_o, name=None):
        _filtered = copy.copy(self.filtered)

        if name:
            pe_o.strat_name = name
            pe_o.strat_allname = name
            pe_o.日pe.策略名 = name
            pe_o.日pe.策略全名 = name
            pe_o.单笔pe.策略名 = name
            pe_o.单笔pe.策略全名 = name
        self.filtered.append(pe_o)
        print(f'添加策略:{name}, 添加前策略数量:{len(_filtered)}, 添加后策略数量:{len(self.filtered)}')

    def add_strat_by_days_pnl(self, days_pnl, name='add_from_days_pnl', 初始资金=1000000, date_from='20000101', 单次滑点单位元=0,
                              date_to='20990101'):
        # if not isinstance(days_pnl, list):
        #     days_pnl = [days_pnl]

        _filtered = self.filtered

        _p = Performance(days_pnl, strat_name=name, strat_allname=name, startTime=date_from,
                         endTime=date_to, 初始资金=初始资金, 单次滑点单位元=单次滑点单位元, )
        self.filtered.append(_p)
        print(f'名字{name}添加前策略数量:{len(self.filtered) - len(_filtered)}, 添加后策略数量:{len(self.filtered)}')
        # print('删掉的策略:\n')

    def trade_time_diff(self, strat_name, time=10, strats_pe='filtered'):
        """
        strats='filtered'
        strats='all_pe_list'
        """
        strats = getattr(self, strats_pe)
        base_strats = [v for v in strats if strat_name not in v.strat_name ]
        print(f"计算 {time} 秒内的交易重复率, 策略:{strat_name} 与 base_strats: {[n.strat_name for n in base_strats]}")
        # pprint.pprint([n.strat_name for n in strats])

        return b.datas.functions.trade_time_diff(b.datas.functions.concat_tradelist([v.tradelist for v in base_strats]),
                                                 self.select_strat_by_name_1(strat_name).tradelist, time)

    def all_strat_trade_time_diff(self, time=10, strats_name='concat', strats_pe='filtered'):
        """
        strats_name='concat_'
        strats_name='all_name'

        strats_pe='filtered'
        strats_pe='all_pe_list'
        """
        datas = []
        strats = getattr(self, strats_name)()
        for name in strats:
            data = self.trade_time_diff(name, time=time, strats_pe=strats_pe)
            datas.append(data)

        self.duplicate_percentage_df = pd.DataFrame(
            data=datas, index=strats,
            columns=['重复占比', '重复占比值', '当前时间平均差异', '所有时间平均差异']
        )
        self.duplicate_percentage_df = self.duplicate_percentage_df.sort_values('重复占比值')
        return self.duplicate_percentage_df

    def list_pe_to_df(self, attribute='日pe', listname=None):
        if not listname:
            listname = self.filtered
        return pd.DataFrame([getattr(o, attribute) for o in listname])

    def filtered_pe(self, 策略总数量=None, 初始资金=None, date_from=None, date_to=None, 单次滑点单位元=0, single_plot=False, name=None, pe_from="filtered"):
        '''和from_df_list的单次滑点单位元参数单独生效,合并交易后的滑点是单独用交易次数计算的,所以互不相干,不会叠加生效'''
        if not 策略总数量:
            策略总数量 = len(self.filtered)

        if not 初始资金:
            初始资金 = self.初始资金 * len(self.filtered)

        print(f"策略总数量:{策略总数量}")
        print(f"初始资金:{初始资金}")

        if single_plot:
            # 生成单个策略日收益矩阵
            self.filtered_pnl_day_single_strat = pd.concat(
                [getattr(v, '日收益回撤df')[date_from:date_to]['日净利润'] for v in self.filtered], axis=1, )
            self.filtered_pnl_day_single_strat.columns = [getattr(strat, 'strat_name') for strat in self.filtered]
            self.filtered_mdd_day_single_strat = pd.concat(
                [getattr(v, '日收益回撤df')[date_from:date_to]['日累计最大回撤金额'] for v in self.filtered], axis=1, )
            self.filtered_mdd_day_single_strat.columns = [getattr(strat, 'strat_name') for strat in self.filtered]

            dfall = self.filtered_pnl_day_single_strat / 10000
            dfmmdall = self.filtered_mdd_day_single_strat / 10000
            dfall.cumsum().iplot(title='单个策略盈亏比较', yTitle='单位：万')  # ,dimensions=(600, 300)
            dfmmdall.iplot(title='单个策略回撤比较', yTitle='单位：万')  # ,dimensions=(600, 300)

            df_year = dfall.groupby(pd.Grouper(freq="1Y")).cumsum()
            df_year.iplot(title='单个策略年度盈亏比较', yTitle='单位：万')  # ,dimensions=(600, 300)

            # df_year = dfall.resample('Y').sum()
            # df_year.index = df_year.index.to_period()
        _pe_from = getattr(self,pe_from)
        all_trade = pd.concat([pe.tradedf for pe in _pe_from])
        self.pe = Performance(all_trade, 初始资金=初始资金, startTime=date_from, endTime=date_to, 单次滑点单位元=单次滑点单位元, )
        if name:
            self.pe.strat_name = name
            self.pe.strat_allname = name
        # return all_trade
        return self.pe
    def strat_rotation(self, 策略总数量=None, 初始资金=None, 做多后n名次=0, 做多前n名次=0, ):
        if not 策略总数量:
            策略总数量 = len(self.filtered)
        if not 初始资金:
            if 做多后n名次:
                初始资金 = 做多后n名次 * self.初始资金
            if 做多前n名次:
                初始资金 = 做多前n名次 * self.初始资金
        print(f"策略总数量:{策略总数量}")
        print(f"初始资金:{初始资金}")
        # if 做多后n名次 >= len(self.filtered):
        #     raise ValueError('轮动个数不能大于等于符合收益率策略个数')
        日收益回撤df_list = [o.日收益回撤df for o in self.filtered]
        # data = {'日期': pd.date_range('2015', pd.datetime.now().date(), freq='b')}
        mdd_dict = {}
        day_pnl_dict = {}
        # data = pd.DataFrame()
        for o in self.filtered:
            # print(o.strat_allname)
            mdd_dict[o.strat_allname] = o.日收益回撤df.日累计最大回撤比率.resample('B').ffill()  # 按工作日重采样，向前填充
            day_pnl_dict[o.strat_allname] = o.日收益回撤df.日净利润.resample('B').asfreq(0)  # 按工作日重采样，向前填充
        df_mdd = pd.DataFrame(mdd_dict)
        df_mdd_sort = df_mdd.copy()
        df_day_pnl = pd.DataFrame(day_pnl_dict)
        dfi = df_mdd.iterrows()
        # 行排序，并按列名写回df
        for i in dfi:
            ss = i[1].sort_values(ascending=False)
            ss = ss.to_frame().assign(s=range(1, len(ss) + 1))
            for j in i[1].index:
                df_mdd_sort.loc[i[0]][j] = int(ss.loc[j].s)

        # return df_mdd  ,df_mdd_sort, df_day_pnl

        if 做多后n名次:
            df_day_pnl[df_mdd_sort.shift(1) <= 策略总数量 - 做多后n名次] = 0
            df_day_pnl.iloc[0] = 0
            return Performance(日收益=df_day_pnl.sum(axis=1), 初始资金=初始资金)
        if 做多前n名次:
            df_day_pnl[df_mdd_sort.shift(1) > 做多前n名次] = 0
            df_day_pnl.iloc[0] = 0
            return Performance(日收益=df_day_pnl.sum(axis=1), 初始资金=初始资金)




# 日内

    def get_feature_within_the_day(self,feature_inf,split_time):
        '''
        (只处理比较简单的形式,复杂的手动获取)
        获取日间和日内特征(可能需要两个函数分别获取,日间和之间有差异)
        feature_inf:[[日内数据],[日间数据]]

        日间数据:使用get_filter_data函数(传入stratege_type(日间为xx_within),index)

        '''
        data = []
        feature_inf_within = feature_inf[0]
        if len(feature_inf_within) > 0:
            data_within_day = self.get_filter_data(feature_inf_within[0],feature_inf_within[1])
            data.append(data_within_day)

        feature_inf_between = feature_inf[1]
        if len(feature_inf_between) > 0:
            data_between_day = self.get_filter_data(feature_inf_between[0],feature_inf_between[1])
            data.append(data_between_day)
        if len(data) > 1:
            r_data = pd.concat(data,axis=1,join='inner')
        else:
            r_data = data[0]

        return r_data

    def get_filter_pe_by_split_time(self,origin_pe,split_time):
        '''
        删除split_time之前的交易,合成新的pe
        '''
        df = origin_pe.单笔收益trade_df.copy()
        df['日内时间'] = df['开仓时间'].apply(lambda x :str(x)[11:])
        dp = df[df['日内时间'] >= split_time].iloc[:,:-1]
        new_pe = pe(单笔收益 = dp,初始资金=origin_pe.初始资金, startTime=origin_pe.startTime, endTime=origin_pe.endTime,strat_allname=f"{origin_pe.strat_allname}_after_{split_time}", strat_name=f"{origin_pe.strat_allname}_after_{split_time}")

        return new_pe


    def strat_increase_pnl_within_the_day(self,origin_pe,feature_data,split_time,func_base = 'get_filter_inf_by_decisiontree',data_period = [100,100],return_all = False,signal_list = [0,1]):
        '''
        日内增强:通过日内(也可以加上日间信息),在日内取一个临界值,通过之前的数据预测之后是否需要交易
        参数:
            origin_pe:需要增强的策略pe
            feature_data:直接传入feature,格式:dataframe(在外部算好)如果需要单因子网格搜索需传入feature_inf            split_time:日内策略划分时间
            func_base:构建增强策略的方法
            data_period:滑动窗口周期
            return_all:True时返回包含split_time前的交易,False时不包含
        return:
            一个新的增强后的pel
        '''
        # 获取特征
        self.strat_increase_pnl_cal_feature_within(split_time = split_time)
        self.strat_increase_pnl_cal_feature()
        if isinstance(feature_data, pd.DataFrame) or isinstance(feature_data, pd.Series):
            feature = feature_data
        else:
            index = feature_data[1] + '_' + feature_data[2]
            feature = self.get_filter_data(strategy_type = feature_data[0],index = index)

        # 根据split_time生成过滤后的新的pe(需要考虑是否要返回之前的交易结果特征)
        new_pe = self.get_filter_pe_by_split_time(origin_pe,split_time)
        print(1)
        # 构建数据集
        data = pd.concat([feature,new_pe.日收益回撤df['日收益率(单利)']],axis=1,join = 'inner')

        # 利用func生成筛选后的pe,分单因子和多因子
        func = getattr(self,func_base)

        if func_base == 'get_filter_inf_by_grid':
            # 单因子网格情况
            print(data)
            filter_inf_list = func(data,data_period,feature_data)
            print(filter_inf_list)
            df = self.strat_increase_pnl_filter_feature_to_signal(filter_inf_list,signal_list = signal_list)
            print('a')
            df_signal = self.concat_signal_pe(origin_pe.日收益回撤df,df)

        else:
            # 多因子算法情况
            signal = func(data,data_period)
            df_signal = self.concat_signal_pe(origin_pe.日收益回撤df,signal)
        print(2)
        if return_all == False:
            # 返回不包含之前split_time之前的pe
            return_pe = self.strat_increase_pnl_filter_to_signal_to_pe(new_pe,df_signal)[0]
            print(3)
            return return_pe
        else:
            # 返回包括之前的pe(split_time之前的交易存在)可能需要一个新的合成函数
            pass

    def strat_increase_pnl_cal_feature_within(self,split_time = '10',if_index = ['000300.XSHG','IF88'],ic_index = ['000905.XSHG','IC88'],ih_index = [],period = '1m',reget = False):
        # 比起日间的需要多一个split_time
        # ic  指数   etf
        now_time = datetime.datetime.now().strftime('%Y-%m-%d')

        if reget == False:
            if getattr(self, 'increase_pnl_cal_feature_within_day',False):
                if getattr(self, 'increase_pnl_cal_feature_within_day').get("data_date") == now_time:
                    return self.increase_pnl_cal_feature_within_day

        def get_data(index,split_time,start_date='2014-01-01', end_date=now_time,period = '1m'):
            '''
            获取日间相关的数据,计算相关指标(需要改动)
            '''
            data = rqdata.get_price(index,start_date=start_date, end_date=end_date, frequency='1m',adjust_type='post')
            data = data.xs(index, level='order_book_id')
            data = data.reset_index()

            data['time'] = data['datetime'].apply(lambda x :str(x)[11:])
            data['date'] = data['datetime'].apply(lambda x :str(x)[:10])
            f_data = data[(data['time'] >= '09:31:00') & (data['time'] < split_time)]

            f_data['highm_lowm'] = f_data['high'] / f_data['low'] - 1
            feature = pd.DataFrame(f_data.groupby('date').agg({'high':'max','low':'min','highm_lowm':'sum'}))
            # 每分钟波动和
            feature[f'{index}_highm_lowm'] = feature['highm_lowm']
            # split前的high - low
            feature[f'{index}_high_low'] = feature['high'] / feature['low'] - 1
            feature['日期'] = pd.to_datetime(feature.index)
            feature = feature.set_index('日期')
            # 后续自行添加,并将数据加入col中

            col = [f'{index}_high_low',f'{index}_highm_lowm']
            print(f'{index}:success')
            return feature[col]


        # index_list = ['000300.XSHG','IF888','000905.XSHG','IC888']

        def get_index_df(index_list):
            if len(index_list) == 0:
                return None
            else:
                data_list = []
                for index in index_list:
                    data_list.append(get_data(index,split_time,period = period))
                if len(data_list) == 1:
                    return data_list[0]
                else:
                    return pd.concat(data_list,axis=1)

        ic_df = get_index_df(ic_index)

        if_df = get_index_df(if_index)

        ih_df = get_index_df(ih_index)

        print("数据:")
        if len(ic_index) > 0:
            print('ic:')
            print(ic_df.columns)
        if len(if_index) > 0:
            print('if:')
            print(if_df.columns)
        if len(ih_index) > 0:
            print('ih:')
            print(ih_df.columns)

        print(f"time:{now_time}")

        # outer_data用来储存利用算法生成的其他数据
        self.increase_pnl_cal_feature_within_day = {"ic_within":ic_df,
                                         "if_within":if_df,
                                         "ih_within":ih_df,
                                         "data_date":now_time,
                                         "outer_data":None}

        return self.increase_pnl_cal_feature_within_day






# 日间

    def concat_signal_pe(self,df_pe,df_all):
        df_signal = pd.concat([df_pe,df_all],axis=1,join='inner')
        return df_signal

    def strat_increase_pnl_cal_feature(self,if_index = ['000300.XSHG','IF88'],ic_index = ['000905.XSHG','IC88'],ih_index = [],period = '1d',reget = False):
        # ic  指数   etf
        now_time = datetime.datetime.now().strftime('%Y-%m-%d')

        if reget == False:
            if getattr(self, 'increase_pnl_cal_feature',False):
                if getattr(self, 'increase_pnl_cal_feature').get("data_date") == now_time:
                    return self.increase_pnl_cal_feature

        def get_data(index,start_date='2014-01-01', end_date=now_time,period = '1d'):
            '''
            获取相关的数据
            '''
            df = rqdata.get_price(index,start_date=start_date, end_date=end_date, frequency=period,adjust_type='post')
            print(f'{index}数据个数:',len(df))
            df['close1'] = df['close'].shift(1)
            # 开盘波动
            df[f'{index}_open_close1'] = df['open'] / df['close1'] - 1
            df['close2'] = df['close'].shift(2)
            # 昨日涨跌
            df[f'{index}_close1_close2'] = df['close1'] / df['close2'] - 1
            # 昨日high-low
            df[f'{index}_high1_low1'] = df['high'].shift(1) / df['low'].shift(1) - 1
            # 其他指标(可自行添加,return时也需添加在col中)


            col = [f'{index}_open_close1',f'{index}_close1_close2',f'{index}_high1_low1']
            print(f'{index}:success')
            return df[col].xs(index, level='order_book_id')

        # index_list = ['000300.XSHG','IF888','000905.XSHG','IC888']

        def get_index_df(index_list):
            if len(index_list) == 0:
                return None
            else:
                data_list = []
                for index in index_list:
                    data_list.append(get_data(index,period = period))
                if len(data_list) == 1:
                    return data_list[0]
                else:
                    return pd.concat(data_list,axis=1)

        ic_df = get_index_df(ic_index)

        if_df = get_index_df(if_index)

        ih_df = get_index_df(ih_index)

        print("数据:")
        if len(ic_index) > 0:
            print('ic:')
            print(ic_df.columns)
        if len(if_index) > 0:
            print('if:')
            print(if_df.columns)
        if len(ih_index) > 0:
            print('ih:')
            print(ih_df.columns)

        print(f"time:{now_time}")

        # outer_data用来储存利用算法生成的其他数据
        self.increase_pnl_cal_feature = {"ic":ic_df,
                                         "if":if_df,
                                         "ih":ih_df,
                                         "data_date":now_time,
                                         "outer_data":None}

        return self.increase_pnl_cal_feature



    def get_filter_data(self,strategy_type,index):
        # index为all时返回strategy_type下全部数据
        if len(strategy_type) < 5:
            self.strat_increase_pnl_cal_feature()
            data = self.increase_pnl_cal_feature[strategy_type]
        else:
            self.strat_increase_pnl_cal_feature_within()
            data = self.increase_pnl_cal_feature_within_day[strategy_type]
        return data[index] if index != 'all' else data




    def strat_increase_pnl_filter_to_signal_to_pe(self,origin_pe,df):
        self.strat_increase_pnl_filter_all_feature_to_signal
        new_pes = []
        for signal_name in df.columns:
            if "_signal" in signal_name:
                new_pes.append(pe(日收益=df.日净利润 * df[signal_name] , 初始资金=origin_pe.初始资金, startTime=origin_pe.startTime, endTime=origin_pe.endTime, strat_allname=f"{origin_pe.strat_allname}_{signal_name}", strat_name=f"{origin_pe.strat_allname}_{signal_name}",))

        return new_pes



    def split_rolling_data(self,data,data_period):
        '''
        将数据划分成为滑动窗口模式
        '''
        train_period = data_period[0]
        test_period = data_period[1]
        # for

        start = train_period
        length = len(data)
        # print(length)
        for i in range(start,length,test_period):
            train_data = data.iloc[i-train_period:i,:]
            test_data = data.iloc[i:min(i+test_period,length),:]
            yield (train_data, test_data)



    def get_filter_inf_by_grid(self,data,data_period ,feature_inf):
        '''
        单因子
        传入data,其中包含单个feature和收益率因变量,data_period为以100天训练预测后100天
        返回筛选条件filter_inf_list(形式和之前一致)
        '''
        def get_single_filter_inf_by_grid(train,test):
            '''
            传入每段数据,通过网格得到每段数据的筛选信息,返回筛选信息
            '''
            # print(1)
            start_time = test.index[0]
            end_time = test.index[-1]
            filter_range = []

            # 上涨边界
            d_max = train.iloc[:,0].max()
            d_range = d_max / 25
            range_high = [d_range * i for i in range(25)]
            range_low = [d_range * i for i in range(20)]
            params = {'range_high':range_high,'range_low':range_low}

            model = GridClass()
            gs = GridSearchCV(model, params)
            gs.fit(train)
            # print('最好参数：{}，最好分数{}'.format(gs.best_params_, gs.best_score_))

            # high_range = [gs.best_params_['range_low'],gs.best_params_['range_high']]

            filter_range.append([gs.best_params_['range_low'],gs.best_params_['range_high'],start_time,end_time])

            # 下跌边界(存在小于0的情况才有下跌筛选边界)
            if train.iloc[:,0].min() < 0:
                d_max = train.iloc[:,0].min()
                d_range = d_max / 25
                range_high = [d_range * i for i in range(25)]
                range_low = [d_range * i for i in range(20)]
                params = {'range_high':range_high,'range_low':range_low}

                model = GridClass()
                gs = GridSearchCV(model, params)
                gs.fit(train)
                # print('最好参数：{}，最好分数{}'.format(gs.best_params_, gs.best_score_))

                filter_range.append([gs.best_params_['range_low'],gs.best_params_['range_high'],start_time,end_time])

            return filter_range

            # return []

        from collections import namedtuple
        filter_inf = namedtuple('filter_inf',['data_type','index_code','condition','filter_range'])
        # s1 = filter_inf('ic','000905.XSHG','close1_close2',[[-0.01,0],[-5,-3]])
        filter_range_list = []
        for train,test in self.split_rolling_data(data,data_period):
            filter_range = get_single_filter_inf_by_grid(train,test)
            filter_range_list += filter_range

        s1 = filter_inf(feature_inf[0],feature_inf[1],feature_inf[2],filter_range_list)

        return [s1]



    def get_filter_inf_by_decisiontree(self,data,data_period):
        '''
        通过决策树选取,思路与网格类似,
        传入的data为dataframe,最后1列为label前面为特征
        返回的内容不是筛选条件,而是预测结果
        '''
        def yield_test(y):
            p = list(range(len(y)))
            yield [p,p]
        def my_score_by_mean_return(y,y_pred,*,beta):
            y_pred = y_pred.reshape(beta.shape)
            # print(3)
            yy = beta.values * y_pred
            # print(4)
            mean_re = yy.sum() / len(yy[yy != 0])

            # print('mean_re',mean_re)
            return mean_re

        def get_single_filter_inf_by_decisiontree(train,test):
            '''
            滑动窗口分段数据
            '''

            col = train.shape[1]
            max_depth = [i for i in range(1,col)]
            max_leaf = [i for i in range(2,col**2)]
            params = {'max_depth':max_depth,'max_leaf_nodes':max_leaf}

            from sklearn.tree import DecisionTreeClassifier
            model = DecisionTreeClassifier()
            cv = yield_test(train)
            label = train.columns[-1]
            t_c = train.copy()
            train['label'] = train[label].apply(lambda x:0 if x < train[label].mean() else 1)

            # 自定义的score
            # cv = yield_test(train)
            # score = make_scorer(my_score_by_mean_return,beta = t_c.iloc[:,-1])

            cv = 5
            score = 'roc_auc'
            gs = GridSearchCV(model, params,scoring=score,cv = cv)

            gs.fit(t_c.iloc[:,:-1],train.iloc[:,-1])
            print('最好参数：{}，最好分数{}'.format(gs.best_params_, gs.best_score_))

            model_use = DecisionTreeClassifier(max_depth = gs.best_params_['max_depth'],max_leaf_nodes=gs.best_params_['max_leaf_nodes'])
            model_use.fit(t_c.iloc[:,:-1],train.iloc[:,-1])
            y_pre = model_use.predict(test.iloc[:,:-1])
            print(model_use.predict(t_c.iloc[:,:-1])[:20])
            print(y_pre[:20])

            test['decisiontree_classify_signal'] = y_pre
            return test


        signal_list = []
        for train,test in self.split_rolling_data(data,data_period):
            siganl = get_single_filter_inf_by_decisiontree(train,test)
            signal_list.append(siganl)

        signal_df = pd.concat(signal_list,axis = 0)
        return signal_df



    def strat_increase_pnl_filter_rolling_feature_to_single_pe_signal(self,origin_pe,feature_inf,data_period = [100,100],signal_list = [0,1],func_base = 'get_filter_inf_by_grid'):
        '''
        传入需要的pe得到因变量,需要的feature_inf得到自变量,筛选的时间间隔和作用间隔,signal对应的权重,需要用来筛选的函数方法
        返回一个包含signal的df

        '''
        label = origin_pe.日收益回撤df['日收益率(单利)']
        index = feature_inf[1] + '_' + feature_inf[2]
        feature = self.get_filter_data(strategy_type = feature_inf[0],index = index)

        data = pd.concat([feature,label],axis = 1,join='inner')
        data = data[~(data['日收益率(单利)'].isna())]

        func = getattr(self,func_base)
        filter_inf_list = func(data,data_period,feature_inf)

        df = self.strat_increase_pnl_filter_feature_to_signal(filter_inf_list,signal_list = signal_list)
        # new_pe_list = self.strat_increase_pnl_filter_to_signal_to_pe(origin_pe,df_signal)
        df_signal = self.concat_signal_pe(origin_pe.日收益回撤df,df)

        return df_signal




    def strat_increase_pnl_filter_feature_to_signal(self,filter_inf_list,signal_list = [0,1]):
        self.strat_increase_pnl_cal_feature()

        df_list = []
        for filter_inf in filter_inf_list:
            # namedtuple('data_type','index_code','condition','filter_range')
            strategy_type = filter_inf.data_type
            index = filter_inf.index_code + '_' + filter_inf.condition

            df = self.get_filter_data(strategy_type,index)
            df = pd.DataFrame(df)
            df[f'{index}_signal'] = signal_list[0]
            for t in filter_inf.filter_range:
                if len(t) == 1:
                    t.append(t[0])
                if len(t) == 2:
                    df.loc[(df[f'{index}'] >= t[0]) & (df[f'{index}'] <= t[1]),f'{index}_signal']= signal_list[1]
                else:
                    df.loc[(df[f'{index}'] >= t[0]) & (df[f'{index}'] <= t[1])& (df.index >= t[2]) & (df.index <= t[3]),f'{index}_signal']= signal_list[1]

            df_list.append(df)
        if len(df_list) == 1:
            df_all = df_list[0]
        else:
            df_all = pd.concat(df_list,axis=1,join='outer')

        self.strat_increase_pnl_filter_all_feature_to_signal = df_all
        return df_all




    def strat_increase_pnl_filter_all(self, filter_inf_list,pe_list = 'filtered',signal_list = [0,1]):
        if not getattr(self,"increase_pnl_cal_feature",False):
            self.strat_increase_pnl_cal_feature()

        if pe_list == 'filtered':
            pe_list = getattr(self,'filtered')

        if not getattr(self,"strat_increase_pnl_filter_all_new_pe",False):
            self.strat_increase_pnl_filter_all_new_pe = []

        self.strat_increase_pnl_filter_feature_to_signal(filter_inf_list,signal_list=signal_list)
        df_all = self.strat_increase_pnl_filter_all_feature_to_signal

        for origin_pe in pe_list:
            df_signal = pd.concat([origin_pe.日收益回撤df,df_all],axis=1,join='inner')

            new_pe_list = self.strat_increase_pnl_filter_to_signal_to_pe(origin_pe,df_signal)

            self.strat_increase_pnl_filter_all_new_pe += new_pe_list

        self.strat_increase_pnl_filter_all_pe = pe_list + self.strat_increase_pnl_filter_all_new_pe

        return self.strat_increase_pnl_filter_all_new_pe






    def corr(self):
        self.daily_rate_of_return = pd.concat([strat.日收益回撤df['日收益率(单利)'] for strat in self.filtered], axis=1)  # 日净利润
        self.daily_rate_of_return.columns = [strat.strat_name for strat in self.filtered]
        return self.daily_rate_of_return.dropna().corr()

    def mddm(self, sort_values='初始资金当前回撤率为', ascending=False, listname=None, ):  # multi_drawdown_merge
        if not listname:
            listname = self.filtered

        try:
            # 当存在某个pe没有单笔属性时,将所有pe都按没有单笔属性进行计算
            for o in listname:
                if getattr(o, '单笔pe') == None:
                    raise Exception("部分pe无单笔属性")
            日pe_all = pd.DataFrame([getattr(o, '日pe') for o in listname]).set_index('策略全名').T
            单笔pe_all = pd.DataFrame([getattr(o, '单笔pe') for o in listname]).set_index('策略全名').T
            # 添加每年的夏普和年化, 通过匹配的方式, 不然无法确定从多少年到多少年;
            日pe_all_rowname = 日pe_all.index.to_list()
            sharpe_ratio_list = sorted([x for x in 日pe_all_rowname if re.search("\'sharpe_ratio\'", str(x))])
            annual_return_list = sorted([x for x in 日pe_all_rowname if re.search("\'年化\'", str(x))])
            dddf = pd.concat([
                日pe_all.loc[
                    ['初始资金年化收益', '初始资金最大回撤率为', '初始资金当前回撤率为', 'sharpe_ratio_v4', '周胜率', '月胜率', '策略交易日', '开盘交易日']
                + sharpe_ratio_list + annual_return_list],
                单笔pe_all.loc[['总交易笔数', '交易日日平均交易笔数', '总滑点金额']]
            ]).T  # .drop('平均').applymap(lambda x:sort_by_float float(str(x).replace('%', ''))
        except:
            日pe_all = pd.DataFrame([getattr(o, '日pe') for o in listname]).set_index('策略全名').T
            单笔pe_all = None
            dddf = pd.concat([日pe_all.loc[['初始资金年化收益', '初始资金最大回撤率为', '初始资金当前回撤率为', 'sharpe_ratio_v4', '周胜率', '月胜率', '策略交易日', '开盘交易日']]]).T

            # def multi_drawdown_merge(日pe_all, 单笔pe_all):


        dddf.insert(0, '起始时间', 日pe_all.loc[['起始时间']].T)
        dddf.insert(1, '结束时间', 日pe_all.loc[['结束时间']].T)
        dddf.index.name = '策略名'
        # dddf.sort_values('初始资金当前回撤率为', ascending=False, inplace=True)
        dddf = b.datas.functions.sort_by_float(df=dddf, col=sort_values, ascending=ascending, )
        return dddf



'''

    tradelist = [
        concat_tradelist([btfm('wang_DL_IH_20_1_20_0_0_2000_par_val_净盈_127_12_1.3_F4'),
                          'Lwang_DL_IH_20_1_20_0_0_5000_par_val_净盈_127_12_1.3_F4', ]),
        concat_tradelist([btfm('QC_dlx_IH_15_1_20_0_0_2000_par_val_净盈_127_12_1.3_F4')[:'2020-9-1'],
                          'LQC_dlx_IH_15_1_20_0_0_5000_par_val_净盈_127_12_1.3_F4', ]),
        concat_tradelist([btfm('wb4g_sp_IH_10_1_20_0_0_2000_par_val_净盈_期指一级趋势线30min_12')[:],
                          'Lwb4g_sp_IH_10_1_20_0_0_5000_par_val_净盈_期指一级趋势线30min_12']),
        concat_tradelist([btfm('wang_DL_IC_20_1_20_0_0_2000_par_val_净盈_127_12_1.3_F4'),
                          'Lwang_DL_IC_20_1_20_0_0_5000_par_val_净盈_127_12_1.3_F4', ]),
        # 4    71.27
        concat_tradelist([btfm('Qzh_IC_32_1_20_0_0_2000_par_val_净盈_127_12_1.3_F4_d')[:'2020-07-03'],
                          'LQzh_IC_32_1_20_0_0_5000_par_val_净盈_127_12_1.3_F4', ]),
        # 5    73.72
        concat_tradelist([btfm('QC_dlx_IC_15_1_20_0_0_2000_par_val_净盈_127_12_1.3_F4')[:'2020-9-1'],
                          'LQC_dlx_IC_15_1_20_0_0_5000_par_val_净盈_127_12_1.3_F4', ]),
        # 'wb4g_IC_10_1_20_0_0_2000_par_val_净盈_期指一级趋势线30min_12_limit1.1',
        # 'wb4g_IC_10_1_20_0_0_2000_par_val_净盈_期指一级趋势线30min_12_limit1.3',
        # 'wb4g_IC_10_1_20_0_0_2000_par_val_净盈_期指一级趋势线30min_12_limit1.5',
        # 'wb4g_IF_10_1_20_0_0_2000_par_val_净盈_期指一级趋势线30min_12_limit1.1',
        # 'wb4g_IF_10_1_20_0_0_2000_par_val_净盈_期指一级趋势线30min_12_limit1.4',

        #     'wb4g_sp_IF_1_1_12_0_0_2000_par_val_净盈_a',
    ]

    # r = Performances().from_df_list(tradelist=eval(b.s.c.s25_01), date_from='2015-8-1', date_to='2030-11-29', )  # mean_pnl=False,  sum_pnl=False)
    # r.filter_by_pnl(30)
    # r.all_strat_pe().p4()
    # r.strat_rotation(4)
    
    '''

'''
    dfall, 单笔pe_all, 日pe_all, compare, corr = cmbt([
        concat_tradelist([btfm('wb4g_sp_IC_6_1_20_0_0_2000_par_val_净盈_期指一级趋势线30min_12#日交易次数_16')[:'2019-11-12'],
                          'Lwb4g_sp_IC_6_1_20_0_0_5000_par_val_净盈_期指一级趋势线30min_12#日交易次数_16', ]),
        concat_tradelist([
            'wb4g_sp_IC_10_1_20_0_0_2000_par_val_净盈_期指一级趋势线30min_12#日交易次数_12',
            'Lwb4g_sp_IC_10_1_20_0_0_39000_par_val_净盈_期指一级趋势线30min_12#日交易次数_12'
        ]),
        concat_tradelist([
            'wb4g_sp_IC_10_1_20_0_0_2000_par_val_净盈_期指一级趋势线30min_12',
            'Lwb4g_sp_IC_10_1_20_0_0_39000_par_val_净盈_期指一级趋势线30min_12'
        ]),
        # 'wb4g_IC_10_1_20_0_0_2000_par_val_净盈_期指一级趋势线30min_12_limit1.1',
        # 'wb4g_IC_10_1_20_0_0_2000_par_val_净盈_期指一级趋势线30min_12_limit1.3',
        # 'wb4g_IC_10_1_20_0_0_2000_par_val_净盈_期指一级趋势线30min_12_limit1.5',
        # 'wb4g_IF_10_1_20_0_0_2000_par_val_净盈_期指一级趋势线30min_12_limit1.1',
        # 'wb4g_IF_10_1_20_0_0_2000_par_val_净盈_期指一级趋势线30min_12_limit1.4',

        #     'wb4g_sp_IF_1_1_12_0_0_2000_par_val_净盈_a',
    ], date_from='2014-01-1', date_to='2030-11-29', sum_pnl=False)  # mean_pnl=False,  sum_pnl=False


    # compare
    pe(dfall.多策略平均 * 10000, 初始资金=1000000, startTime='2014-01-1', endTime='2030-11-29').p4()
    '''

'''
日初始资金 = 初始资金
df_day['日累计净利润'] = df_day['日净利润'].cumsum()
df_day['日权益'] = df_day['日累计净利润'] + 日初始资金
df_day['期初权益'] = df_day['日累计净利润'] - df_day['日净利润'] + 日初始资金
df_day['日累计收益率'] = df_day['日累计净利润'] / 日初始资金
df_day['日收益率'] = (df_day['日净利润'] / df_day['日权益'].shift(1))
df_day['日最大权益'] = df_day['日权益'].expanding(min_periods=1).max()  # 计算当日之前的账户最大价值
df_day['日累计最大回撤金额'] = df_day['日权益'] - df_day['日最大权益']
df_day['日初始资金最大回撤比率'] = df_day['日权益'] / df_day['日最大权益'] - 1  # 计算当日的回撤


# df.insert(0, '日期', df['平仓时间'].str[:10])
# df.insert(0, '单笔累计利润', df['净利润'].cumsum())
df = df.groupby('日期').sum()
df['累计利润']=df['净利润'].cumsum()
df['权益']=df['累计利润']+initial_cash
df['日收益率'] = (df['净利润']/df['权益'].shift(1))
df['对数日收益率'] = (df['权益']/df['权益'].shift(1)).map(np.log)
df['还原对数日收益率'] = df['对数日收益率'].map(np.exp) -1
df['累计对数日收益率'] = df['对数日收益率'] .cumsum()
df['还原累计对数日收益率'] = df['累计对数日收益率'].map(np.exp)
df['日超额回报率'] = df['日收益率'] - 0.04/252
df['夏普比率'] = math.sqrt(252)* df['日超额回报率'].mean()/df['日超额回报率'].std()
df['累计收益率'] = np.cumprod(1 + df['日收益率'])
df['累计超额收益率'] = np.cumprod(1+df['日超额回报率'])-1
df = df.round(4)
# df.to_csv(self.logFolder+'tradeLog'+self.logFile)   
    
    # # beta值
# beta(date_line, return_line, indexreturn_line)
# # alpha值
# alpha(date_line, capital_line, index_line, return_line, indexreturn_line)\
# # 信息比率
# info_ratio(date_line, return_line, indexreturn_line)
# 我将介绍如下几个经典指标
年化收益
任何收益都应该转换成年化收益，以进行比较。股神巴菲特投资几十年以来的年化收益是20%左右。美国股市百年来年化收益大概是10%左右。所以我们也不要太贪心，不要动辄就要求每年翻倍的收益。计算公式为：（账户最终价值/账户初始价值）^（250/回测期间总天数）-1

最大回撤
最大回撤就是从一个高点到一个低点最大的下跌幅度，用来描述我们的策略可能出现的最糟糕的情况，衡量了最极端可能的亏损。例如一个策略的最大回撤是50%，那么你使用这个策略之前就要掂量掂量，自己是否能经受得起50%的下跌。计算公式为：min(账户当日价值 / 当日之前账户最高价值-1)。

我一般认为，年华收益/最大回撤，是个最简单有效的策略判断指标。

平均涨幅
平均涨幅衡量了在回测期间资产的平均涨跌情况。计算公式为：账户日收益的平均值。

上涨概率
上涨概率衡量了我们投资组合上涨的可能性。计算公式为：上涨天数 / 回测交易日数量。

最大连续上涨天数
衡量了我们策略连续盈利的最大天数。

最大连续下跌天数
衡量了我们策略连续亏损的最大天数。或者说策略连续失效的次数。这个指标很重要，在实战中，若连续失败的次数多了，会动摇自己的信心，使自己很难跟着策略走下去。

最大单周期涨幅
所有日期中最大的涨幅，衡量了资产一天内的最好表现。

最大单周期跌幅
所有日期中最大的跌幅，衡量了资产一天内的最差表现。

（【python量化课程】想要快速、系统的学习量化知识，可以参与我与论坛合作开设的课程：《python量化投资入门》，我会亲自授课，随问随答。参与课程还可以免费加入我的小密圈，我每天会在圈中分享量化的所见所思，圈子介绍点击此处。）

收益波动率
单单追求高收益率是不行的，我们在投资中还必须考虑到潜在的风险。波动太大的策略往往风险也较高。收益波动率就衡量了策略收益的波动情况即风险。计算公式为：账户日收益的年化标准差。

贝塔(beta)值
根据经典的资本资产定价理论(CAPM模型)，β系数衡量了资产的回报率对市场变动的敏感程度，代表了该资产的系统性风险，表示策略对大盘的敏感性。计算公式为：账户日收益与参考基准日收益的协方差 / 参考基准日收益的方差。

阿尔法(alpha)值
虽然我们的策略会受到大盘的影响，但是每个策略都会有自己市场因素之外的收益，alpha值表示实际风险回报和平均预期风险回报的差额，衡量了投资的非系统性风险。计算公式为：（账户年化收益-无风险收益）-beta*（参考基准年化收益-无风险收益）。

夏普比率
理性的投资者将选择并持有有效的投资组合，即那些在给定的风险水平下使期望回报最大的投资组合，夏普比率就是一个可以同时对收益与风险加以综合考虑的经典指标，表示每承受一单位风险，会产生多少的超额回报。计算公式为：（账户年化收益率-无风险利率）/ 收益波动率。

信息比率
信息比率以马克维茨的均异模型为基础，用来衡量超额风险所带来的超额收益。它表示单位主动风险所带来的超额收益。计算公式为：（账户日收益 - 参考基准日收益）的年化均值/年化标准差。
'''
from sklearn.base import BaseEstimator, ClassifierMixin
from sklearn.metrics import make_scorer
from sklearn.model_selection import GridSearchCV
# import hmmlearn

class GridClass(BaseEstimator):
    """An example of classifier"""

    def __init__(self,range_high =0.1, range_low = 0):
        """
        Called when initializing the classifier
        """
        # import pdb; pdb.set_trace()
        # self.params = params
        # if params != None:
        #     range_high = self.params['range_high']
        #     range_low = params['range_low']

        self.range_high = range_high
        self.range_low = range_low

        # THIS IS WRONG! Parameters should have same name as attributes
        # self.differentParam = otherParam

    def fit(self, data, y=None):
        """
        This should fit classifier. All the "work" should be done here.

        Note: assert is not a good choice here and you should rather
        use try/except blog with exceptions. This is just for short syntax.
        """
        self.x = data.iloc[:,0]
        self.y = data.iloc[:,1]
        self.mean = self.y.mean()
        index = data.columns[0]
        # print(index)
        self.filter_data = data[(data[index] > self.range_low) & (data[index] <= self.range_high)]
        # self.filter_high_data = data[(data[index] > self.range_high[0]) & (data[index] <= self.range_high[1])]

        return self

    def _meaning(self, x):
        # returns True/False according to fitted classifier
        # notice underscore on the beginning
        pass

    def predict(self, data, y=None):
        return data[(data.iloc[:,0] > self.range_low) & (data.iloc[:,0] <= self.range_high)]

    def score(self, X, y=None):
        # counts number of values bigger than mean
        # self.filter_data.iloc[:,1].sum()
        # import pdb; pdb.set_trace()

        # print(self.filter_data.iloc[:,1].mean() if len(self.filter_data) > len(self.x) / 4 else 0)
        return self.filter_data.iloc[:,1].mean() if len(self.filter_data) > len(self.x) / 4 else 0

```

```python
class TurtleTradingStrategy(bt.Strategy):
    params = dict(
        N1= 20, # 唐奇安通道上轨的t
        N2=10, # 唐奇安通道下轨的t
        )
    
    def log(self, txt, dt=None):           
            dt = dt or self.datas[0].datetime.date(0)
            print('%s, %s' % (dt.isoformat(), txt))

    def __init__(self): 
        self.order = None                   
        self.buy_count = 0 # 记录买入次数
        self.last_price = 0 # 记录买入价格
        # 准备第一个标的沪深300主力合约的close、high、low 行情数据
        self.close = self.datas[0].close
        self.high = self.datas[0].high
        self.low = self.datas[0].low
        # 计算唐奇安通道上轨：过去20日的最高价
        self.DonchianH = bt.ind.Highest(self.high(-1), period=self.p.N1, subplot=True)
        # 计算唐奇安通道下轨：过去10日的最低价
        self.DonchianL = bt.ind.Lowest(self.low(-1), period=self.p.N2, subplot=True)
        # 生成唐奇安通道上轨突破：close>DonchianH，取值为1.0；反之为 -1.0
        self.CrossoverH = bt.ind.CrossOver(self.close(0), self.DonchianH, subplot=False)
        # 生成唐奇安通道下轨突破:
        self.CrossoverL = bt.ind.CrossOver(self.close(0), self.DonchianL, subplot=False)
        # 计算 ATR
        self.TR = bt.ind.Max((self.high(0)-self.low(0)), # 当日最高价-当日最低价
                                    abs(self.high(0)-self.close(-1)), # abs(当日最高价−前一日收盘价)
                                    abs(self.low(0)-self.close(-1))) # abs(当日最低价-前一日收盘价)
        self.ATR = bt.ind.SimpleMovingAverage(self.TR, period=self.p.N1, subplot=False)
        # 计算 ATR，直接调用 talib ，使用前需要安装 python3 -m pip install TA-Lib
        # self.ATR = bt.talib.ATR(self.high, self.low, self.close, timeperiod=self.p.N1, subplot=True)
    
    
    def next(self): 
        # 如果还有订单在执行中，就不做新的仓位调整
        if self.order:
            return  
                
        # 如果当前持有多单
        if self.position.size > 0 :
            # 多单加仓:价格上涨了买入价的0.5的ATR且加仓次数少于等于3次
            if self.datas[0].close >self.last_price + 0.5*self.ATR[0] and self.buy_count <= 4:
                print('if self.datas[0].close >self.last_price + 0.5*self.ATR[0] and self.buy_count <= 4:')
                print('self.buy_count',self.buy_count)
                # 计算建仓单位：self.ATR*期货合约乘数300*保证金比例0.1
                self.buy_unit = max((self.broker.getvalue()*0.005)/(self.ATR*300*0.1),1)
                self.buy_unit = int(self.buy_unit) # 交易单位为手
                # self.sizer.p.stake = self.buy_unit
                self.order = self.buy(size=self.buy_unit)
                self.last_price = self.position.price # 获取买入价格
                self.buy_count = self.buy_count + 1
            #多单止损：当价格回落2倍ATR时止损平仓
            elif self.datas[0].close < (self.last_price - 2*self.ATR[0]):
                print('elif self.datas[0].close < (self.last_price - 2*self.ATR[0]):')
                self.order = self.sell(size=abs(self.position.size))
                self.buy_count = 0
            #多单止盈：当价格突破10日最低点时止盈离场 平仓
            elif self.CrossoverL < 0:
                print('self.CrossoverL < 0')
                self.order = self.sell(size=abs(self.position.size))
                self.buy_count = 0 
                
        # 如果当前持有空单
        elif self.position.size < 0 :
            # 空单加仓:价格小于买入价的0.5的ATR且加仓次数少于等于3次
            if self.datas[0].close<self.last_price-0.5*self.ATR[0] and self.buy_count <= 4:
                print('self.datas[0].close<self.last_price-0.5*self.ATR[0] and self.buy_count <= 4')
                # 计算建仓单位：self.ATR*期货合约乘数300*保证金比例0.1
                self.buy_unit = max((self.broker.getvalue()*0.005)/(self.ATR*300*0.1),1)
                self.buy_unit = int(self.buy_unit) # 交易单位为手
                # self.sizer.p.stake = self.buy_unit
                self.order = self.sell(size=self.buy_unit)
                self.last_price = self.position.price # 获取买入价格
                self.buy_count = self.buy_count + 1              
            #空单止损：当价格上涨至2倍ATR时止损平仓
            elif self.datas[0].close < (self.last_price+2*self.ATR[0]):
                print('self.datas[0].close < (self.last_price+2*self.ATR[0])')
                self.order = self.buy(size=abs(self.position.size))
                self.buy_count = 0
            #多单止盈：当价格突破20日最高点时止盈平仓
            elif self.CrossoverH>0:
                print('self.CrossoverH>0')
                self.order = self.buy(size=abs(self.position.size))
                self.buy_count = 0
                
        else: # 如果没有持仓，等待入场时机
            #入场: 价格突破上轨线且空仓时，做多
            if self.CrossoverH > 0 and self.buy_count == 0:
                print('if self.CrossoverH > 0 and self.buy_count == 0:')
                # 计算建仓单位：self.ATR*期货合约乘数300*保证金比例0.1
                self.buy_unit = max((self.broker.getvalue()*0.005)/(self.ATR*300*0.1),1)
                self.buy_unit = int(self.buy_unit) # 交易单位为手
                self.order = self.buy(size=self.buy_unit)
                self.last_price = self.position.price # 记录买入价格
                self.buy_count = 1  # 记录本次交易价格
            #入场: 价格跌破下轨线且空仓时，做空
            elif self.CrossoverL < 0 and self.buy_count == 0:
                print('self.CrossoverL < 0 and self.buy_count == 0')
                # 计算建仓单位：self.ATR*期货合约乘数300*保证金比例0.1
                self.buy_unit = max((self.broker.getvalue()*0.005)/(self.ATR*300*0.1),1)
                self.buy_unit = int(self.buy_unit) # 交易单位为手
                self.order = self.sell(size=self.buy_unit)
                self.last_price = self.position.price # 记录买入价格
                self.buy_count = 1  # 记录本次交易价格
        
    # 打印订单日志
    def notify_order(self, order):
        order_status = ['Created','Submitted','Accepted','Partial',
                        'Completed','Canceled','Expired','Margin','Rejected']
        # 未被处理的订单
        if order.status in [order.Submitted, order.Accepted]:
            self.log('ref:%.0f, name: %s, Order: %s'% (order.ref,
                                                   order.data._name,
                                                   order_status[order.status]))
            return
        # 已经处理的订单
        if order.status in [order.Partial, order.Completed]:
            if order.isbuy():
                self.log(
                        'BUY EXECUTED, status: %s, ref:%.0f, name: %s, Size: %.2f, Price: %.2f, Cost: %.2f, Comm %.2f' %
                        (order_status[order.status], # 订单状态
                         order.ref, # 订单编号
                         order.data._name, # 股票名称
                         order.executed.size, # 成交量
                         order.executed.price, # 成交价
                         order.executed.value, # 成交额
                         order.executed.comm)) # 佣金
            else: # Sell
                self.log('SELL EXECUTED, status: %s, ref:%.0f, name: %s, Size: %.2f, Price: %.2f, Cost: %.2f, Comm %.2f' %
                            (order_status[order.status],
                             order.ref,
                             order.data._name,
                             order.executed.size,
                             order.executed.price,
                             order.executed.value,
                             order.executed.comm))
                    
        elif order.status in [order.Canceled, order.Margin, order.Rejected, order.Expired]:
            # 订单未完成
            self.log('ref:%.0f, name: %s, status: %s'% (
                order.ref, order.data._name, order_status[order.status]))
            
        self.order = None
        
    def notify_trade(self, trade):
        # 交易刚打开时
        if trade.justopened:
            self.log('Trade Opened, name: %s, Size: %.2f,Price: %.2f' % (
                    trade.getdataname(), trade.size, trade.price))
        # 交易结束
        elif trade.isclosed:
            self.log('Trade Closed, name: %s, GROSS %.2f, NET %.2f, Comm %.2f' %(
            trade.getdataname(), trade.pnl, trade.pnlcomm, trade.commission))
        # 更新交易状态
        else:
            self.log('Trade Updated, name: %s, Size: %.2f,Price: %.2f' % (
                    trade.getdataname(), trade.size, trade.price))
   
  
```
